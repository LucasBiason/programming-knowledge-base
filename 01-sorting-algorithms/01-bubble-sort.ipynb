# Bubble Sort

Objetivo: Ordenar uma lista comparando elementos adjacentes e trocando-os quando estão fora de ordem.

Complexidade:
- Tempo: O(n²) no pior caso, O(n) no melhor caso
- Espaço: O(1)

Categoria: Ordenação

## Fundamentos

Bubble sort é o algoritmo de ordenação mais simples de entender. Compara elementos adjacentes e troca-os se estiverem na ordem errada. O processo repete até não haver mais trocas necessárias.

O nome vem do comportamento: os maiores elementos "sobem" para o final da lista, como bolhas subindo na água. A cada passagem completa, pelo menos um elemento vai para sua posição final correta.

É um dos primeiros algoritmos ensinados em cursos de ciência da computação por sua simplicidade, apesar da baixa eficiência em listas grandes.


```python
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

def bubble_sort_animated(arr):
    """
    Gera animacao passo a passo do Bubble Sort.
    Cada frame mostra uma comparacao, com cores indicando o estado de cada elemento.
    """
    arr = arr.copy()
    n = len(arr)
    frames = []

    # Frame inicial
    frames.append({
        "arr": arr.copy(),
        "comparing": [],
        "sorted_positions": [],
        "label": "Array original"
    })

    for i in range(n):
        swapped = False
        for j in range(0, n - 1 - i):
            # Frame: comparando j e j+1
            frames.append({
                "arr": arr.copy(),
                "comparing": [j, j + 1],
                "sorted_positions": list(range(n - i, n)),
                "label": f"Comparando [{j}]={arr[j]} e [{j+1}]={arr[j+1]}"
            })

            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
                # Frame: apos troca
                frames.append({
                    "arr": arr.copy(),
                    "comparing": [j, j + 1],
                    "sorted_positions": list(range(n - i, n)),
                    "label": f"Troca! [{j}]={arr[j]} <-> [{j+1}]={arr[j+1]}"
                })

        # Frame: fim da passagem
        frames.append({
            "arr": arr.copy(),
            "comparing": [],
            "sorted_positions": list(range(n - i - 1, n)),
            "label": f"Passagem {i+1} completa - posicao [{n-1-i}] definida"
        })

        if not swapped:
            break

    # Frame final
    frames.append({
        "arr": arr.copy(),
        "comparing": [],
        "sorted_positions": list(range(n)),
        "label": "Vetor totalmente ordenado"
    })

    return frames


def animate_bubble_sort(arr):
    """
    Renderiza a animacao do Bubble Sort no Jupyter notebook.
    """
    frames = bubble_sort_animated(arr)

    fig, ax = plt.subplots(figsize=(10, 4))
    max_val = max(arr) * 1.3

    def draw_frame(frame_idx):
        ax.clear()
        frame = frames[frame_idx]
        data = frame["arr"]
        n = len(data)
        colors = []

        for idx in range(n):
            if idx in frame["comparing"]:
                colors.append("#f59e0b")    # amarelo - comparando
            elif idx in frame["sorted_positions"]:
                colors.append("#10b981")    # verde - posicao final
            else:
                colors.append("#3b82f6")    # azul - normal

        bars = ax.bar(range(n), data, color=colors, edgecolor="black", linewidth=1.5)

        for idx, (bar, val) in enumerate(zip(bars, data)):
            ax.text(
                bar.get_x() + bar.get_width() / 2.0,
                bar.get_height() + max_val * 0.02,
                str(val),
                ha="center", va="bottom", fontsize=14, fontweight="bold"
            )

        ax.set_xlim(-0.5, n - 0.5)
        ax.set_ylim(0, max_val)
        ax.set_xticks(range(n))
        ax.set_xticklabels([f"[{i}]" for i in range(n)], fontsize=11)
        ax.set_ylabel("Valor", fontsize=11)
        ax.set_title(
            f"Bubble Sort  |  Frame {frame_idx + 1}/{len(frames)}",
            fontsize=13, fontweight="bold", pad=12
        )

        ax.text(
            0.5, -0.12, frame["label"],
            transform=ax.transAxes, ha="center", fontsize=12,
            style="italic", color="#374151"
        )

        legend_elements = [
            mpatches.Patch(color="#3b82f6", label="Normal"),
            mpatches.Patch(color="#f59e0b", label="Comparando"),
            mpatches.Patch(color="#10b981", label="Posicao final"),
        ]
        ax.legend(handles=legend_elements, loc="upper right", fontsize=9)
        ax.grid(axis="y", alpha=0.3, linestyle="--")
        fig.subplots_adjust(bottom=0.18)

    anim = FuncAnimation(
        fig, draw_frame,
        frames=len(frames),
        interval=800,
        repeat=True
    )
    plt.close(fig)
    return HTML(anim.to_jshtml())


# Executar animacao com o array de exemplo
animate_bubble_sort([5, 3, 8, 4, 2])
```

## Como Funciona

1. Percorre a lista do início ao fim
2. Compara cada elemento com seu vizinho da direita
3. Se o elemento da esquerda for maior, troca-os
4. Repete o processo até não haver mais trocas

Otimização: usar uma flag para detectar se houve trocas na passagem atual. Se não houve trocas, a lista já está ordenada e podemos parar.


## Análise de Complexidade

### Tempo

Pior Caso: O(n²) — Ocorre quando a lista está inversamente ordenada. Cada elemento precisa ser movido até sua posição final, resultando em n passagens com n comparações cada.

Caso Médio: O(n²) — Mesmo comportamento do pior caso na maioria dos cenários.

Melhor Caso: O(n) — Quando a lista já está ordenada. Com a otimização da flag, apenas uma passagem é necessária para confirmar que não há trocas.

### Espaço

O(1) — O algoritmo ordena in-place, usando apenas variáveis auxiliares para trocas. Não requer espaço adicional proporcional ao tamanho da entrada.

### Estabilidade

Sim — elementos com valores iguais mantêm sua ordem relativa original.


```python
def bubble_sort(arr):
    """
    Sorts an array using bubble sort algorithm.
    
    Args:
        arr: List of comparable elements
        
    Returns:
        Sorted list (modified in-place)
        
    Time: O(n²) worst case, O(n) best case
    Space: O(1)
    """
    n = len(arr)
    
    # Outer loop: number of passes needed
    for i in range(n):
        swapped = False
        
        # Inner loop: compare adjacent elements
        # After each pass, largest element is in place
        # So we can reduce the range by i each time
        for j in range(0, n - 1 - i):
            if arr[j] > arr[j + 1]:
                # Swap elements
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        
        # Optimization: if no swaps, list is sorted
        if not swapped:
            break
    
    return arr
```

## Detalhes da Implementação

- O loop externo controla quantas passagens completas fazemos pela lista. O loop interno compara pares adjacentes e troca quando necessário.
- A otimização com a flag `swapped` permite terminar antecipadamente quando a lista já está ordenada. Sem essa otimização, o algoritmo sempre faz n passagens completas, mesmo quando desnecessário.
- A redução do range interno (`n - 1 - i`) é uma segunda otimização. Como a cada passagem o maior elemento vai para o final, não precisamos comparar esses elementos já ordenados nas próximas iterações.


```python
# Test with small array
test_arr = [5, 3, 8, 4, 2]
print(f"Original: {test_arr}")

bubble_sort(test_arr)
print(f"Sorted: {test_arr}")
```

```python
# Edge cases
empty = []
single = [42]
sorted_arr = [1, 2, 3, 4, 5]
reverse = [5, 4, 3, 2, 1]
duplicates = [3, 1, 4, 1, 5, 9, 2, 6, 5]

test_cases = [
    ("Empty", empty),
    ("Single element", single),
    ("Already sorted", sorted_arr),
    ("Reverse sorted", reverse),
    ("With duplicates", duplicates)
]

for name, arr in test_cases:
    original = arr.copy()
    bubble_sort(arr)
    print(f"{name}: {original} -> {arr}")
```

## Aplicações Práticas

### Quando Usar

- Listas muito pequenas (menos de 10 elementos)
- Quando simplicidade é mais importante que performance
- Contextos educacionais para entender ordenação básica

### Quando Evitar

- Listas grandes (mais de 100 elementos)
- Aplicações que exigem alta performance
- Quando a lista pode estar parcialmente ordenada (outros algoritmos são melhores)


## Comparacao

| Algoritmo | Tempo (Pior) | Tempo (Melhor) | Espaco | Estavel | In-Place | Quando Usar |
|-----------|--------------|----------------|--------|---------|----------|-------------|
| Bubble Sort | O(n^2) | O(n) | O(1) | Sim | Sim | Educacional, listas muito pequenas |
| Selection Sort | O(n^2) | O(n^2) | O(1) | Nao | Sim | Quando swaps sao custosos |
| Insertion Sort | O(n^2) | O(n) | O(1) | Sim | Sim | Listas quase ordenadas, pequenas |
| Merge Sort | O(n log n) | O(n log n) | O(n) | Sim | Nao | Garantia de performance, estabilidade |
| Quick Sort | O(n^2) | O(n log n) | O(log n) | Nao | Sim | Caso geral rapido, melhor caso medio |

Bubble sort e o mais simples, mas o mais lento entre os algoritmos basicos. Para listas grandes, algoritmos O(n log n) como merge sort e quick sort sao necessarios.


## Perguntas Comuns em Entrevistas

### Nível Básico

1. Como funciona o bubble sort?
**R:** Compara elementos adjacentes e troca-os se estiverem fora de ordem. Repete até não haver mais trocas.

2. Qual a complexidade do bubble sort?
**R:** O(n²) no pior caso, O(n) no melhor caso (com otimização), O(1) de espaço.

### Nível Intermediário

1. Como otimizar o bubble sort?
**R:** Usar uma flag para detectar se houve trocas. Se não houve trocas em uma passagem, a lista está ordenada e podemos parar. Também reduzir o range interno a cada iteração.

### Nível Avançado

1. Por que bubble sort é considerado ineficiente mesmo com otimizações?
**R:** Mesmo com otimizações, o pior caso continua O(n²). Algoritmos como merge sort e quick sort oferecem O(n log n) garantido ou no caso médio, sendo muito mais eficientes para listas grandes.

## Exercicios

**Desafio 1:** Implemente o Cocktail Sort (variante bidirecional do bubble sort) que percorre a lista alternando entre os sentidos esquerda-direita e direita-esquerda.

**Desafio 2:** Modifique o bubble sort para contar e retornar o numero total de comparacoes realizadas junto com o array ordenado.

```python
# Desafio 1: Cocktail Sort
def cocktail_sort(arr):
    n = len(arr)
    swapped = True
    start = 0
    end = n - 1
    while swapped:
        swapped = False
        for i in range(start, end):
            if arr[i] > arr[i + 1]:
                arr[i], arr[i + 1] = arr[i + 1], arr[i]
                swapped = True
        if not swapped:
            break
        end -= 1
        swapped = False
        for i in range(end - 1, start - 1, -1):
            if arr[i] > arr[i + 1]:
                arr[i], arr[i + 1] = arr[i + 1], arr[i]
                swapped = True
        start += 1
    return arr

# Desafio 2: Bubble sort com contador de comparacoes
def bubble_sort_count(arr):
    n = len(arr)
    comparisons = 0
    for i in range(n):
        swapped = False
        for j in range(0, n - 1 - i):
            comparisons += 1
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        if not swapped:
            break
    return arr, comparisons

# Teste
arr1 = [5, 3, 8, 4, 2]
print("Cocktail Sort:", cocktail_sort(arr1.copy()))

arr2 = [5, 3, 8, 4, 2]
sorted_arr, comps = bubble_sort_count(arr2)
print(f"Bubble com contagem: {sorted_arr} | Comparacoes: {comps}")
```

## Benchmark

Comparacao de desempenho com o `sorted()` nativo do Python para arrays de tamanhos 100, 1.000 e 5.000 elementos.

```python
import timeit
import random

def benchmark_bubble_sort():
    sizes = [100, 1000, 5000]
    print(f"{'Tamanho':>10} | {'bubble_sort (ms)':>18} | {'sorted() (ms)':>15}")
    print("-" * 52)
    for n in sizes:
        data = random.sample(range(n * 10), n)

        t_bubble = timeit.timeit(
            lambda d=data: bubble_sort(d.copy()), number=5
        ) / 5 * 1000

        t_sorted = timeit.timeit(
            lambda d=data: sorted(d), number=5
        ) / 5 * 1000

        print(f"{n:>10} | {t_bubble:>18.4f} | {t_sorted:>15.4f}")

benchmark_bubble_sort()
```
