# Selection Sort

Objetivo: Encontrar o menor elemento e movê-lo para a posição correta, repetindo para o restante da lista.

Complexidade:
- Tempo: O(n²) em todos os casos
- Espaço: O(1)

Categoria: Ordenação

## Fundamentos

Selection sort é o algoritmo que "seleciona" o menor elemento em cada iteração e coloca-o na posição correta. Diferente do bubble sort que faz múltiplas trocas, o selection sort faz apenas uma troca por passagem.

A analogia perfeita é selecionar jogadores para um time: você varre todos os candidatos, encontra o melhor, e coloca-o na primeira posição. Depois repete para a segunda posição, terceira, e assim por diante.

É interessante quando o custo de escrita na memória é alto (swaps custam caro), pois faz exatamente n-1 trocas no total - o mínimo possível para qualquer algoritmo de ordenação por comparação.


```python
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

def selection_sort_animated(arr):
    """
    Gera animacao passo a passo do Selection Sort.
    Cada frame mostra a busca pelo minimo e a troca.
    """
    arr = arr.copy()
    n = len(arr)
    frames = []

    frames.append({
        "arr": arr.copy(),
        "current": [],
        "min_idx": None,
        "sorted_positions": [],
        "label": "Array original"
    })

    for i in range(n - 1):
        min_idx = i

        # Frame: inicio da passagem, posicao i como candidata
        frames.append({
            "arr": arr.copy(),
            "current": [i],
            "min_idx": i,
            "sorted_positions": list(range(i)),
            "label": f"Passagem {i+1}: buscando menor a partir de [{i}]"
        })

        for j in range(i + 1, n):
            # Frame: comparando j com min_idx
            frames.append({
                "arr": arr.copy(),
                "current": [j],
                "min_idx": min_idx,
                "sorted_positions": list(range(i)),
                "label": f"Comparando [{j}]={arr[j]} com minimo [{min_idx}]={arr[min_idx]}"
            })

            if arr[j] < arr[min_idx]:
                min_idx = j
                frames.append({
                    "arr": arr.copy(),
                    "current": [],
                    "min_idx": min_idx,
                    "sorted_positions": list(range(i)),
                    "label": f"Novo minimo encontrado: [{min_idx}]={arr[min_idx]}"
                })

        # Troca
        if min_idx != i:
            arr[i], arr[min_idx] = arr[min_idx], arr[i]
            frames.append({
                "arr": arr.copy(),
                "current": [i, min_idx],
                "min_idx": None,
                "sorted_positions": list(range(i + 1)),
                "label": f"Troca [{i}] <-> [{min_idx}] - posicao [{i}] definida"
            })
        else:
            frames.append({
                "arr": arr.copy(),
                "current": [],
                "min_idx": None,
                "sorted_positions": list(range(i + 1)),
                "label": f"Sem troca - [{i}]={arr[i]} ja e o menor"
            })

    # Frame final
    frames.append({
        "arr": arr.copy(),
        "current": [],
        "min_idx": None,
        "sorted_positions": list(range(n)),
        "label": "Vetor totalmente ordenado"
    })

    return frames


def animate_selection_sort(arr):
    """
    Renderiza a animacao do Selection Sort no Jupyter notebook.
    """
    frames = selection_sort_animated(arr)

    fig, ax = plt.subplots(figsize=(10, 4))
    max_val = max(arr) * 1.3

    def draw_frame(frame_idx):
        ax.clear()
        frame = frames[frame_idx]
        data = frame["arr"]
        n = len(data)
        colors = []

        for idx in range(n):
            if idx in frame["sorted_positions"]:
                colors.append("#10b981")    # verde - posicao final
            elif frame["min_idx"] is not None and idx == frame["min_idx"]:
                colors.append("#ef4444")    # vermelho - minimo atual
            elif idx in frame["current"]:
                colors.append("#f59e0b")    # amarelo - elemento sendo verificado
            else:
                colors.append("#3b82f6")    # azul - normal

        bars = ax.bar(range(n), data, color=colors, edgecolor="black", linewidth=1.5)

        for idx, (bar, val) in enumerate(zip(bars, data)):
            ax.text(
                bar.get_x() + bar.get_width() / 2.0,
                bar.get_height() + max_val * 0.02,
                str(val),
                ha="center", va="bottom", fontsize=14, fontweight="bold"
            )

        ax.set_xlim(-0.5, n - 0.5)
        ax.set_ylim(0, max_val)
        ax.set_xticks(range(n))
        ax.set_xticklabels([f"[{i}]" for i in range(n)], fontsize=11)
        ax.set_ylabel("Valor", fontsize=11)
        ax.set_title(
            f"Selection Sort  |  Frame {frame_idx + 1}/{len(frames)}",
            fontsize=13, fontweight="bold", pad=12
        )

        ax.text(
            0.5, -0.12, frame["label"],
            transform=ax.transAxes, ha="center", fontsize=12,
            style="italic", color="#374151"
        )

        legend_elements = [
            mpatches.Patch(color="#3b82f6", label="Normal"),
            mpatches.Patch(color="#f59e0b", label="Verificando"),
            mpatches.Patch(color="#ef4444", label="Minimo atual"),
            mpatches.Patch(color="#10b981", label="Posicao final"),
        ]
        ax.legend(handles=legend_elements, loc="upper right", fontsize=9)
        ax.grid(axis="y", alpha=0.3, linestyle="--")
        fig.subplots_adjust(bottom=0.18)

    anim = FuncAnimation(
        fig, draw_frame,
        frames=len(frames),
        interval=800,
        repeat=True
    )
    plt.close(fig)
    return HTML(anim.to_jshtml())


# Executar animacao com o array de exemplo
animate_selection_sort([5, 3, 8, 4, 2])
```

## Como Funciona

1. Encontra o menor elemento na parte não ordenada
2. Troca com o primeiro elemento da parte não ordenada
3. Move a fronteira entre ordenado/não-ordenado uma posição à direita
4. Repete até toda a lista estar ordenada

Característica: Faz exatamente n-1 trocas, independente da entrada. Mesmo que a lista já esteja ordenada, o algoritmo continua procurando o mínimo.


## Análise de Complexidade

### Tempo

Pior Caso: O(n²)
Mesmo que a lista esteja ordenada, o algoritmo varre toda a parte não ordenada procurando o mínimo. Sempre faz (n-1) + (n-2) + ... + 1 = n(n-1)/2 comparações.

Caso Médio: O(n²)
Comportamento idêntico ao pior caso.

Melhor Caso: O(n²)
Diferente de outros algoritmos, não há otimização para listas já ordenadas. A complexidade é sempre O(n²).

### Espaço

O(1)
Ordena in-place, usando apenas variáveis auxiliares para armazenar o índice do mínimo. Não requer espaço adicional.

### Estabilidade

Não - a troca de posições pode alterar a ordem relativa de elementos iguais.

Exemplo: [4ₐ, 2, 4ᵦ, 1] → [1, 2, 4ᵦ, 4ₐ] (4ᵦ passou na frente de 4ₐ)


```python
def selection_sort(arr):
    """
    Sorts an array using selection sort algorithm.
    
    Args:
        arr: List of comparable elements
        
    Returns:
        Sorted list (modified in-place)
        
    Time: O(n²) in all cases
    Space: O(1)
    """
    n = len(arr)
    
    # Outer loop: position to fill with minimum element
    for i in range(n - 1):
        # Assume current position has the minimum
        min_idx = i
        
        # Inner loop: find actual minimum in unsorted part
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        
        # Swap only if minimum is not already in position
        if min_idx != i:
            arr[i], arr[min_idx] = arr[min_idx], arr[i]
    
    return arr
```

## Detalhes da Implementação

O loop externo define a posição que será preenchida com o próximo menor elemento. O loop interno varre toda a parte não ordenada procurando o índice do menor elemento.

- A variável `min_idx` armazena a posição do menor elemento encontrado. Ao final de cada varredura, fazemos apenas uma troca, colocando o menor na posição correta.

- A verificação `if min_idx != i` evita trocas desnecessárias quando o menor elemento já está na posição correta. Embora não melhore a complexidade O(n²), reduz o número de escritas na memória.


```python
# Test with small array
test_arr = [29, 10, 14, 37, 13]
print(f"Original: {test_arr}")

selection_sort(test_arr)
print(f"Sorted: {test_arr}")
```

```python
# Edge cases
empty = []
single = [42]
sorted_arr = [1, 2, 3, 4, 5]
reverse = [5, 4, 3, 2, 1]
duplicates = [3, 1, 4, 1, 5, 9, 2, 6, 5]

test_cases = [
    ("Empty", empty),
    ("Single element", single),
    ("Already sorted", sorted_arr),
    ("Reverse sorted", reverse),
    ("With duplicates", duplicates)
]

for name, arr in test_cases:
    original = arr.copy()
    selection_sort(arr)
    print(f"{name}: {original} -> {arr}")
```

## Aplicações Práticas

### Quando Usar

- Quando o custo de escrita na memória é alto (flash memory, EEPROM)
- Listas pequenas onde simplicidade importa
- Contextos educacionais para entender ordenação por seleção

### Quando Evitar

- Listas grandes (sempre O(n²), sem casos melhores)
- Quando a lista já está parcialmente ordenada
- Quando estabilidade é necessária


## Comparacao

| Algoritmo | Tempo (Pior) | Tempo (Melhor) | Espaco | Estavel | In-Place | Quando Usar |
|-----------|--------------|----------------|--------|---------|----------|-------------|
| Bubble Sort | O(n^2) | O(n) | O(1) | Sim | Sim | Educacional, listas muito pequenas |
| Selection Sort | O(n^2) | O(n^2) | O(1) | Nao | Sim | Quando swaps sao custosos |
| Insertion Sort | O(n^2) | O(n) | O(1) | Sim | Sim | Listas quase ordenadas, pequenas |
| Merge Sort | O(n log n) | O(n log n) | O(n) | Sim | Nao | Garantia de performance, estabilidade |
| Quick Sort | O(n^2) | O(n log n) | O(log n) | Nao | Sim | Caso geral rapido, melhor caso medio |

Selection sort faz o menor numero de trocas possivel (n-1). Isso e vantajoso quando operacoes de escrita sao custosas, mas nao compensa a complexidade O(n^2) constante em todos os casos.


## Perguntas Comuns em Entrevistas

### Nível Básico

1. Qual a principal diferença entre selection sort e bubble sort?
**R:** Selection sort faz apenas n-1 trocas (uma por iteração), enquanto bubble sort pode fazer O(n²) trocas. Selection sort sempre é O(n²), bubble sort pode ser O(n) no melhor caso.

2. Por que selection sort não é estável?
**R:** A troca de elementos distantes pode alterar a ordem relativa de elementos iguais. Exemplo: [4ₐ, 2, 4ᵦ, 1] → após trocar 4ₐ com 1, temos [1, 2, 4ᵦ, 4ₐ].

### Nível Intermediário

1. Quando selection sort é melhor que bubble sort?
**R:** Quando o custo de escrita na memória é alto. Selection sort minimiza trocas (n-1 total), enquanto bubble sort pode fazer O(n²) trocas. Em flash memory ou EEPROM, isso faz diferença significativa.

### Nível Avançado

1. É possível tornar selection sort estável? Como?
**R:** Sim, mas perde a vantagem do mínimo de trocas. Em vez de trocar diretamente, deslocamos elementos à direita e inserimos o mínimo (similar ao insertion sort). Isso mantém a ordem relativa mas aumenta operações de memória.

## Exercicios

**Desafio 1:** Implemente uma variante do selection sort que ordena em ordem decrescente.

**Desafio 2:** Implemente uma funcao `kth_smallest(arr, k)` que usa selection sort para encontrar o k-esimo menor elemento sem ordenar todo o array.

```python
# Desafio 1: Selection sort decrescente
def selection_sort_desc(arr):
    n = len(arr)
    for i in range(n - 1):
        max_idx = i
        for j in range(i + 1, n):
            if arr[j] > arr[max_idx]:
                max_idx = j
        if max_idx != i:
            arr[i], arr[max_idx] = arr[max_idx], arr[i]
    return arr

# Desafio 2: K-esimo menor elemento
def kth_smallest(arr, k):
    arr = arr.copy()
    n = len(arr)
    for i in range(k):
        min_idx = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr[k - 1]

# Teste
arr = [29, 10, 14, 37, 13]
print("Decrescente:", selection_sort_desc(arr.copy()))

arr2 = [29, 10, 14, 37, 13]
print(f"3o menor de {arr2}:", kth_smallest(arr2, 3))  # esperado: 14
```

## Benchmark

Comparacao de desempenho com o `sorted()` nativo do Python para arrays de tamanhos 100, 1.000 e 5.000 elementos.

```python
import timeit
import random

def benchmark_selection_sort():
    sizes = [100, 1000, 5000]
    print(f"{'Tamanho':>10} | {'selection_sort (ms)':>20} | {'sorted() (ms)':>15}")
    print("-" * 54)
    for n in sizes:
        data = random.sample(range(n * 10), n)

        t_sel = timeit.timeit(
            lambda d=data: selection_sort(d.copy()), number=5
        ) / 5 * 1000

        t_sorted = timeit.timeit(
            lambda d=data: sorted(d), number=5
        ) / 5 * 1000

        print(f"{n:>10} | {t_sel:>20.4f} | {t_sorted:>15.4f}")

benchmark_selection_sort()
```
