# Merge Sort

Objetivo: Dividir a lista em sublistas menores, ordenar recursivamente e combinar (merge) as partes ordenadas.

Complexidade:
- Tempo: O(n log n) em todos os casos
- Espaço: O(n)

Categoria: Ordenação | Técnica: Divisão e Conquista

## Fundamentos

Merge sort é o primeiro algoritmo "profissional" da lista. Enquanto bubble, selection e insertion são O(n²), o merge sort alcança O(n log n) usando a técnica de Divisão e Conquista.

A analogia perfeita é corrigir provas: em vez de corrigir 100 provas de uma vez, você divide em 2 pilhas de 50, depois em 4 pilhas de 25, até ter pilhas de 1 prova. Depois combina as pilhas já corrigidas de forma ordenada.

Criado por John von Neumann em 1945, é um dos algoritmos mais importantes da computação. É estável, previsível (sempre O(n log n)) e forma a base do Timsort usado no Python.


```python
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

def merge_sort_animated(arr):
    """
    Gera animacao passo a passo do Merge Sort.
    Mostra divisao recursiva e merge das sublistas.
    """
    frames = []
    arr = arr.copy()
    n = len(arr)

    frames.append({
        "arr": arr.copy(),
        "highlight": [],
        "merging": [],
        "done": [],
        "label": "Array original"
    })

    def merge_sort_track(arr, left, right):
        if right - left <= 1:
            return

        mid = (left + right) // 2

        # Frame: mostrando divisao
        frames.append({
            "arr": arr.copy(),
            "highlight": list(range(left, mid)),
            "merging": list(range(mid, right)),
            "done": [],
            "label": f"Dividindo: esquerda [{left}:{mid}] e direita [{mid}:{right}]"
        })

        merge_sort_track(arr, left, mid)
        merge_sort_track(arr, mid, right)

        # Merge
        temp = []
        i, j = left, mid
        while i < mid and j < right:
            if arr[i] <= arr[j]:
                temp.append(arr[i])
                i += 1
            else:
                temp.append(arr[j])
                j += 1
        while i < mid:
            temp.append(arr[i])
            i += 1
        while j < right:
            temp.append(arr[j])
            j += 1

        for k, val in enumerate(temp):
            arr[left + k] = val

        # Frame: resultado do merge
        frames.append({
            "arr": arr.copy(),
            "highlight": [],
            "merging": list(range(left, right)),
            "done": [],
            "label": f"Merge completo: posicoes [{left}:{right}] = {temp}"
        })

    merge_sort_track(arr, 0, n)

    # Frame final
    frames.append({
        "arr": arr.copy(),
        "highlight": [],
        "merging": [],
        "done": list(range(n)),
        "label": "Vetor totalmente ordenado"
    })

    return frames


def animate_merge_sort(arr):
    """
    Renderiza a animacao do Merge Sort no Jupyter notebook.
    """
    frames = merge_sort_animated(arr)

    fig, ax = plt.subplots(figsize=(10, 4))
    max_val = max(arr) * 1.3

    def draw_frame(frame_idx):
        ax.clear()
        frame = frames[frame_idx]
        data = frame["arr"]
        n = len(data)
        colors = []

        for idx in range(n):
            if idx in frame["done"]:
                colors.append("#10b981")    # verde - ordenado final
            elif idx in frame["merging"]:
                colors.append("#f59e0b")    # amarelo - sendo merged
            elif idx in frame["highlight"]:
                colors.append("#8b5cf6")    # roxo - sublista esquerda
            else:
                colors.append("#3b82f6")    # azul - normal

        bars = ax.bar(range(n), data, color=colors, edgecolor="black", linewidth=1.5)

        for idx, (bar, val) in enumerate(zip(bars, data)):
            ax.text(
                bar.get_x() + bar.get_width() / 2.0,
                bar.get_height() + max_val * 0.02,
                str(val),
                ha="center", va="bottom", fontsize=14, fontweight="bold"
            )

        ax.set_xlim(-0.5, n - 0.5)
        ax.set_ylim(0, max_val)
        ax.set_xticks(range(n))
        ax.set_xticklabels([f"[{i}]" for i in range(n)], fontsize=11)
        ax.set_ylabel("Valor", fontsize=11)
        ax.set_title(
            f"Merge Sort  |  Frame {frame_idx + 1}/{len(frames)}",
            fontsize=13, fontweight="bold", pad=12
        )

        ax.text(
            0.5, -0.12, frame["label"],
            transform=ax.transAxes, ha="center", fontsize=12,
            style="italic", color="#374151"
        )

        legend_elements = [
            mpatches.Patch(color="#3b82f6", label="Normal"),
            mpatches.Patch(color="#8b5cf6", label="Sublista esquerda"),
            mpatches.Patch(color="#f59e0b", label="Merge em andamento"),
            mpatches.Patch(color="#10b981", label="Ordenado"),
        ]
        ax.legend(handles=legend_elements, loc="upper right", fontsize=9)
        ax.grid(axis="y", alpha=0.3, linestyle="--")
        fig.subplots_adjust(bottom=0.18)

    anim = FuncAnimation(
        fig, draw_frame,
        frames=len(frames),
        interval=1000,
        repeat=True
    )
    plt.close(fig)
    return HTML(anim.to_jshtml())


# Executar animacao com o array de exemplo
animate_merge_sort([5, 3, 8, 4, 2])
```

## Como Funciona

Fase 1: Divisão (Divide)

1. Divide a lista ao meio recursivamente
2. Continua dividindo até ter sublistas de 1 elemento
3. Sublistas de 1 elemento estão automaticamente ordenadas

Fase 2: Conquista (Merge)

4. Combina duas sublistas ordenadas em uma lista maior ordenada
5. Compara os primeiros elementos de cada sublista
6. Insere o menor no resultado
7. Repete até processar todos os elementos

Chave: A fase de merge é O(n) e ocorre em O(log n) níveis da árvore de recursão, resultando em O(n log n) total.


## Análise de Complexidade

### Tempo

Pior Caso: O(n log n) 
Sempre divide a lista ao meio (log n níveis) e faz merge de n elementos em cada nível. Não importa a entrada, a complexidade é sempre a mesma.

Caso Médio: O(n log n)
Comportamento idêntico ao pior caso.

Melhor Caso: O(n log n)
Mesmo para listas já ordenadas, o algoritmo divide e faz merge. Não há otimização para casos melhores.

### Espaço

O(n)
Requer array auxiliar temporário para o merge. Esse é o "preço" pago pela garantia de O(n log n).

### Estabilidade

Sim - durante o merge, quando elementos são iguais, pegamos primeiro da sublista esquerda, mantendo a ordem original.


```python
def merge_sort(arr):
    """
    Sorts an array using merge sort algorithm.
    
    Args:
        arr: List of comparable elements
        
    Returns:
        Sorted list (new list created)
        
    Time: O(n log n) in all cases
    Space: O(n)
    """
    # Base case: list with 0 or 1 element is already sorted
    if len(arr) <= 1:
        return arr
    
    # Divide: split list in half
    mid = len(arr) // 2
    left = arr[:mid]
    right = arr[mid:]
    
    # Conquer: recursively sort both halves
    left = merge_sort(left)
    right = merge_sort(right)
    
    # Combine: merge sorted halves
    return merge(left, right)


def merge(left, right):
    """
    Merges two sorted lists into one sorted list.
    
    Args:
        left: Sorted list
        right: Sorted list
        
    Returns:
        Merged sorted list
    """
    result = []
    i = j = 0
    
    # Compare elements from both lists and add smaller to result
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:  # <= ensures stability
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    
    # Add remaining elements (one list will be empty)
    result.extend(left[i:])
    result.extend(right[j:])
    
    return result
```

## Detalhes da Implementação

A função `merge_sort` divide recursivamente até ter sublistas de 1 elemento (caso base). A recursão cria uma árvore de chamadas com altura log n.

A função `merge` é onde acontece a "mágica". Ela combina duas listas ordenadas em uma lista maior ordenada, comparando sempre os primeiros elementos de cada lista.

A condição `left[i] <= right[j]` (e não `<`) garante estabilidade: quando elementos são iguais, pegamos primeiro da lista esquerda, preservando a ordem original.


```python
# Test with small array
test_arr = [38, 27, 43, 3, 9, 82, 10]
print(f"Original: {test_arr}")

sorted_arr = merge_sort(test_arr)
print(f"Sorted: {sorted_arr}")
```

```python
# Edge cases
empty = []
single = [42]
sorted_arr = [1, 2, 3, 4, 5]
reverse = [5, 4, 3, 2, 1]
duplicates = [3, 1, 4, 1, 5, 9, 2, 6, 5]

test_cases = [
    ("Empty", empty),
    ("Single element", single),
    ("Already sorted", sorted_arr),
    ("Reverse sorted", reverse),
    ("With duplicates", duplicates)
]

for name, arr in test_cases:
    sorted_result = merge_sort(arr)
    print(f"{name}: {arr} -> {sorted_result}")
```

## Aplicações Práticas

### Quando Usar

- Listas grandes onde O(n log n) garantido é necessário
- Quando estabilidade é crucial (ordenação de objetos complexos)
- Ordenação de linked lists (não requer acesso aleatório)
- Ordenação externa (dados não cabem na memória)

### Quando Evitar

- Quando memória é limitada (requer O(n) espaço extra)
- Quando quick sort é viável (geralmente mais rápido na prática)
- Listas muito pequenas (overhead de recursão não compensa)


## Comparacao

| Algoritmo | Tempo (Pior) | Tempo (Melhor) | Espaco | Estavel | In-Place | Quando Usar |
|-----------|--------------|----------------|--------|---------|----------|-------------|
| Bubble Sort | O(n^2) | O(n) | O(1) | Sim | Sim | Educacional, listas muito pequenas |
| Selection Sort | O(n^2) | O(n^2) | O(1) | Nao | Sim | Quando swaps sao custosos |
| Insertion Sort | O(n^2) | O(n) | O(1) | Sim | Sim | Listas quase ordenadas, pequenas |
| Merge Sort | O(n log n) | O(n log n) | O(n) | Sim | Nao | Garantia de performance, estabilidade |
| Quick Sort | O(n^2) | O(n log n) | O(log n) | Nao | Sim | Caso geral rapido, melhor caso medio |

Merge sort e o unico algoritmo aqui com O(n log n) garantido em todos os casos. O preco e O(n) de espaco extra. Quick sort e mais rapido na pratica, mas pode degradar para O(n^2).


## Perguntas Comuns em Entrevistas

### Nível Básico

1. O que é "Divisão e Conquista"?
**R:** Técnica que divide o problema em subproblemas menores, resolve cada um recursivamente e combina as soluções. No merge sort: divide em metades, ordena recursivamente, faz merge.

2. Por que merge sort é O(n log n)?
**R:** A árvore de recursão tem altura log n (dividindo ao meio) e cada nível processa n elementos no merge. Total: n × log n = O(n log n).

### Nível Intermediário

1. Como implementar merge sort in-place para economizar memória?
**R:** É muito complexo e perde a vantagem de simplicidade. Existem versões in-place mas são raramente usadas. Na prática, se memória é crítica, usa-se heap sort (O(n log n) com O(1) espaço).

### Nível Avançado

1. Como adaptar merge sort para ordenação externa (dados em disco)?
**R:** External merge sort: divide dados em chunks que cabem na memória, ordena cada chunk, depois faz k-way merge dos chunks ordenados lendo do disco. Usado em bancos de dados para ordenar tabelas gigantes.

## Exercicios

**Desafio 1:** Implemente o Bottom-Up Merge Sort (iterativo), que evita recursao e ordena subarray de tamanho crescente (1, 2, 4, 8...).

**Desafio 2:** Implemente uma funcao que conta o numero de inversoes em um array usando merge sort (par (i,j) onde i < j mas arr[i] > arr[j]).

```python
# Desafio 1: Bottom-Up Merge Sort (iterativo)
def merge_bottom_up(left, right):
    result = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i]); i += 1
        else:
            result.append(right[j]); j += 1
    result.extend(left[i:])
    result.extend(right[j:])
    return result

def merge_sort_bottom_up(arr):
    n = len(arr)
    result = arr.copy()
    size = 1
    while size < n:
        for start in range(0, n, size * 2):
            mid = min(start + size, n)
            end = min(start + size * 2, n)
            merged = merge_bottom_up(result[start:mid], result[mid:end])
            result[start:end] = merged
        size *= 2
    return result

# Desafio 2: Contar inversoes com merge sort
def count_inversions(arr):
    if len(arr) <= 1:
        return arr, 0
    mid = len(arr) // 2
    left, left_inv = count_inversions(arr[:mid])
    right, right_inv = count_inversions(arr[mid:])
    merged = []
    inversions = left_inv + right_inv
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            merged.append(left[i]); i += 1
        else:
            merged.append(right[j]); j += 1
            inversions += len(left) - i  # todos os restantes em left sao inversoes
    merged.extend(left[i:])
    merged.extend(right[j:])
    return merged, inversions

# Teste
arr = [38, 27, 43, 3, 9, 82, 10]
print("Bottom-Up Merge Sort:", merge_sort_bottom_up(arr))

arr2 = [3, 1, 2, 5, 4]
sorted_arr, inv = count_inversions(arr2)
print(f"Inversoes em {arr2}: {inv} | Ordenado: {sorted_arr}")
```

## Benchmark

Comparacao de desempenho com o `sorted()` nativo do Python para arrays de tamanhos 100, 1.000 e 5.000 elementos.

```python
import timeit
import random

def benchmark_merge_sort():
    sizes = [100, 1000, 5000]
    print(f"{'Tamanho':>10} | {'merge_sort (ms)':>17} | {'sorted() (ms)':>15}")
    print("-" * 50)
    for n in sizes:
        data = random.sample(range(n * 10), n)

        t_merge = timeit.timeit(
            lambda d=data: merge_sort(d.copy()), number=5
        ) / 5 * 1000

        t_sorted = timeit.timeit(
            lambda d=data: sorted(d), number=5
        ) / 5 * 1000

        print(f"{n:>10} | {t_merge:>17.4f} | {t_sorted:>15.4f}")

benchmark_merge_sort()
```
