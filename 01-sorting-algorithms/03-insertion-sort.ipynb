# Insertion Sort

Objetivo: Construir a lista ordenada incrementalmente, inserindo cada elemento na posição correta.

Complexidade:
- Tempo: O(n²) no pior caso, O(n) no melhor caso
- Espaço: O(1)

Categoria: Ordenação

## Fundamentos

Insertion sort é o algoritmo mais "intuitivo" dos três básicos. É exatamente como a maioria das pessoas organiza cartas de baralho na mão: pega uma carta nova e insere na posição correta entre as cartas já ordenadas.

A analogia do baralho é perfeita: sua mão esquerda segura cartas já ordenadas, sua mão direita pega uma nova carta da mesa, você desliza as cartas maiores para a direita e insere a nova carta no espaço criado.

É o algoritmo usado por algoritmos industriais como Timsort (Python/Java) para ordenar sublistas pequenas, pois para listas pequenas ou quase ordenadas, ele supera até o quick sort devido à baixa sobrecarga.


```python
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

def insertion_sort_animated(arr):
    """
    Gera animacao passo a passo do Insertion Sort.
    Mostra a chave sendo inserida na posicao correta da parte ordenada.
    """
    arr = arr.copy()
    n = len(arr)
    frames = []

    frames.append({
        "arr": arr.copy(),
        "key_idx": None,
        "shifting": [],
        "sorted_boundary": 1,
        "label": "Array original - posicao [0] ja e considerada ordenada"
    })

    for i in range(1, n):
        key = arr[i]

        # Frame: selecionar chave
        frames.append({
            "arr": arr.copy(),
            "key_idx": i,
            "shifting": [],
            "sorted_boundary": i,
            "label": f"Chave selecionada: [{i}]={key} - inserir na parte ordenada"
        })

        j = i - 1
        while j >= 0 and arr[j] > key:
            # Frame: deslocando elemento pra direita
            arr[j + 1] = arr[j]
            frames.append({
                "arr": arr.copy(),
                "key_idx": None,
                "shifting": [j + 1],
                "sorted_boundary": i,
                "label": f"Deslocando [{j}]={arr[j]} para [{j+1}] (abrindo espaco para {key})"
            })
            j -= 1

        # Inserir chave na posicao correta
        arr[j + 1] = key
        frames.append({
            "arr": arr.copy(),
            "key_idx": j + 1,
            "shifting": [],
            "sorted_boundary": i + 1,
            "label": f"Chave {key} inserida na posicao [{j+1}]"
        })

    # Frame final
    frames.append({
        "arr": arr.copy(),
        "key_idx": None,
        "shifting": [],
        "sorted_boundary": n,
        "label": "Vetor totalmente ordenado"
    })

    return frames


def animate_insertion_sort(arr):
    """
    Renderiza a animacao do Insertion Sort no Jupyter notebook.
    """
    frames = insertion_sort_animated(arr)

    fig, ax = plt.subplots(figsize=(10, 4))
    max_val = max(arr) * 1.3

    def draw_frame(frame_idx):
        ax.clear()
        frame = frames[frame_idx]
        data = frame["arr"]
        n = len(data)
        boundary = frame["sorted_boundary"]
        colors = []

        for idx in range(n):
            if frame["key_idx"] is not None and idx == frame["key_idx"]:
                colors.append("#f59e0b")    # amarelo - chave
            elif idx in frame["shifting"]:
                colors.append("#ef4444")    # vermelho - deslocando
            elif idx < boundary:
                colors.append("#10b981")    # verde - parte ordenada
            else:
                colors.append("#3b82f6")    # azul - nao processado

        # No frame final, tudo verde
        if frame["label"] == "Vetor totalmente ordenado":
            colors = ["#10b981"] * n

        bars = ax.bar(range(n), data, color=colors, edgecolor="black", linewidth=1.5)

        for idx, (bar, val) in enumerate(zip(bars, data)):
            ax.text(
                bar.get_x() + bar.get_width() / 2.0,
                bar.get_height() + max_val * 0.02,
                str(val),
                ha="center", va="bottom", fontsize=14, fontweight="bold"
            )

        ax.set_xlim(-0.5, n - 0.5)
        ax.set_ylim(0, max_val)
        ax.set_xticks(range(n))
        ax.set_xticklabels([f"[{i}]" for i in range(n)], fontsize=11)
        ax.set_ylabel("Valor", fontsize=11)
        ax.set_title(
            f"Insertion Sort  |  Frame {frame_idx + 1}/{len(frames)}",
            fontsize=13, fontweight="bold", pad=12
        )

        ax.text(
            0.5, -0.12, frame["label"],
            transform=ax.transAxes, ha="center", fontsize=12,
            style="italic", color="#374151"
        )

        legend_elements = [
            mpatches.Patch(color="#3b82f6", label="Nao processado"),
            mpatches.Patch(color="#f59e0b", label="Chave"),
            mpatches.Patch(color="#ef4444", label="Deslocando"),
            mpatches.Patch(color="#10b981", label="Parte ordenada"),
        ]
        ax.legend(handles=legend_elements, loc="upper right", fontsize=9)
        ax.grid(axis="y", alpha=0.3, linestyle="--")
        fig.subplots_adjust(bottom=0.18)

    anim = FuncAnimation(
        fig, draw_frame,
        frames=len(frames),
        interval=800,
        repeat=True
    )
    plt.close(fig)
    return HTML(anim.to_jshtml())


# Executar animacao com o array de exemplo
animate_insertion_sort([5, 3, 8, 4, 2])
```

## Como Funciona

1. Assume que o primeiro elemento já está "ordenado" (lista de 1 elemento)
2. Pega o próximo elemento (chave) da parte não ordenada
3. Compara com elementos à esquerda, da direita para esquerda
4. Desloca elementos maiores uma posição à direita
5. Insere a chave no espaço criado
6. Repete até processar todos os elementos

Diferencial: Usa deslocamento (shift) em vez de trocas (swap). O loop `while` desloca elementos maiores até encontrar a posição correta.


## Análise de Complexidade

### Tempo

Pior Caso: O(n²)
Ocorre quando a lista está inversamente ordenada. Cada elemento precisa ser comparado e deslocado através de toda a parte ordenada.

Caso Médio: O(n²)
Em média, cada elemento precisa ser comparado com metade dos elementos já ordenados.

Melhor Caso: O(n) 
Quando a lista já está ordenada ou quase ordenada. O loop `while` nunca executa, fazendo apenas uma comparação por elemento. Por isso é usado no Timsort!

### Espaço

O(1)
Ordena in-place, usando apenas variáveis auxiliares (chave e índice). Não requer espaço adicional.

### Estabilidade

Sim - elementos iguais mantêm sua ordem relativa porque só deslocamos quando o elemento é estritamente menor.


```python
def insertion_sort(arr):
    """
    Sorts an array using insertion sort algorithm.
    
    Args:
        arr: List of comparable elements
        
    Returns:
        Sorted list (modified in-place)
        
    Time: O(n²) worst case, O(n) best case
    Space: O(1)
    """
    n = len(arr)
    
    # Start from second element (index 1)
    # First element is considered already sorted
    for i in range(1, n):
        # Save the current element (key)
        key = arr[i]
        
        # Start comparing with element to the left
        j = i - 1
        
        # Shift loop: move elements greater than key to the right
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]  # Shift right
            j -= 1
        
        # Insert key in the correct position
        arr[j + 1] = key
    
    return arr
```

## Detalhes da Implementação

O loop externo começa do segundo elemento (índice 1), pois assumimos que o primeiro elemento sozinho já está "ordenado".

A variável `key` guarda o elemento que queremos inserir. O loop `while` desloca elementos maiores uma posição à direita, criando um "espaço" para a chave.

A condição `key < arr[j]` (e não `<=`) garante estabilidade: elementos iguais não são deslocados, mantendo a ordem original.


```python
# Test with small array
test_arr = [5, 2, 4, 6, 1]
print(f"Original: {test_arr}")

insertion_sort(test_arr)
print(f"Sorted: {test_arr}")
```

```python
# Edge cases
empty = []
single = [42]
sorted_arr = [1, 2, 3, 4, 5]
reverse = [5, 4, 3, 2, 1]
duplicates = [3, 1, 4, 1, 5, 9, 2, 6, 5]
almost_sorted = [1, 2, 3, 5, 4]  # Best case scenario

test_cases = [
    ("Empty", empty),
    ("Single element", single),
    ("Already sorted", sorted_arr),
    ("Reverse sorted", reverse),
    ("With duplicates", duplicates),
    ("Almost sorted", almost_sorted)
]

for name, arr in test_cases:
    original = arr.copy()
    insertion_sort(arr)
    print(f"{name}: {original} -> {arr}")
```

## Aplicações Práticas

### Quando Usar

- Listas pequenas (menos de 20 elementos) - mais rápido que quick/merge sort
- Listas quase ordenadas - O(n) de complexidade
- Ordenação online (elementos chegam um por vez)
- Parte do Timsort (Python) para sublistas pequenas

### Quando Evitar

- Listas grandes completamente desordenadas
- Quando há muitos elementos fora de ordem
- Aplicações que precisam de O(n log n) garantido


## Comparacao

| Algoritmo | Tempo (Pior) | Tempo (Melhor) | Espaco | Estavel | In-Place | Quando Usar |
|-----------|--------------|----------------|--------|---------|----------|-------------|
| Bubble Sort | O(n^2) | O(n) | O(1) | Sim | Sim | Educacional, listas muito pequenas |
| Selection Sort | O(n^2) | O(n^2) | O(1) | Nao | Sim | Quando swaps sao custosos |
| Insertion Sort | O(n^2) | O(n) | O(1) | Sim | Sim | Listas quase ordenadas, pequenas |
| Merge Sort | O(n log n) | O(n log n) | O(n) | Sim | Nao | Garantia de performance, estabilidade |
| Quick Sort | O(n^2) | O(n log n) | O(log n) | Nao | Sim | Caso geral rapido, melhor caso medio |

Insertion sort tem o melhor caso O(n), tornando-o ideal para dados quase ordenados. O Timsort usa insertion sort para sublistas pequenas porque muitos dados reais tem runs parcialmente ordenados.


## Perguntas Comuns em Entrevistas

### Nível Básico

1. Por que insertion sort é O(n) no melhor caso?
**R:** Quando a lista já está ordenada, o loop while nunca executa. Apenas fazemos uma comparação por elemento, totalizando O(n).

2. Qual a diferença entre insertion sort e selection sort?
**R:** Insertion sort constrói a parte ordenada inserindo elementos um por um. Selection sort seleciona o menor da parte não ordenada. Insertion é melhor para dados quase ordenados.

### Nível Intermediário

1. Por que o Timsort usa insertion sort?
**R:** Para sublistas pequenas (< 64 elementos), insertion sort é mais rápido que merge sort devido à baixa sobrecarga. Timsort detecta "runs" ordenados e usa insertion sort para completá-los eficientemente.

### Nível Avançado

1. Como adaptar insertion sort para ordenação decrescente mantendo estabilidade?
**R:** Inverter a condição de comparação para `key > arr[j]` no loop while. A estabilidade se mantém porque só deslocamos quando há desigualdade estrita.

## Exercicios

**Desafio 1:** Implemente o Binary Insertion Sort, que usa busca binaria para encontrar a posicao de insercao (reduz comparacoes para O(n log n), mas deslocamentos continuam O(n^2)).

**Desafio 2:** Adapte o insertion sort para ordenar strings ignorando maiusculas e minusculas.

```python
import bisect

# Desafio 1: Binary Insertion Sort
def binary_insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        # Encontra posicao via busca binaria
        pos = bisect.bisect_left(arr, key, 0, i)
        # Desloca elementos
        arr[pos + 1:i + 1] = arr[pos:i]
        arr[pos] = key
    return arr

# Desafio 2: Insertion sort para strings case-insensitive
def insertion_sort_strings(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key.lower() < arr[j].lower():
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr

# Teste
nums = [5, 2, 4, 6, 1, 3]
print("Binary Insertion Sort:", binary_insertion_sort(nums.copy()))

words = ["banana", "Apple", "cherry", "avocado", "Blueberry"]
print("Strings ordenadas:", insertion_sort_strings(words.copy()))
```

## Benchmark

Comparacao de desempenho com o `sorted()` nativo do Python para arrays de tamanhos 100, 1.000 e 5.000 elementos.

```python
import timeit
import random

def benchmark_insertion_sort():
    sizes = [100, 1000, 5000]
    print(f"{'Tamanho':>10} | {'insertion_sort (ms)':>20} | {'sorted() (ms)':>15}")
    print("-" * 54)
    for n in sizes:
        data = random.sample(range(n * 10), n)

        t_ins = timeit.timeit(
            lambda d=data: insertion_sort(d.copy()), number=5
        ) / 5 * 1000

        t_sorted = timeit.timeit(
            lambda d=data: sorted(d), number=5
        ) / 5 * 1000

        print(f"{n:>10} | {t_ins:>20.4f} | {t_sorted:>15.4f}")

benchmark_insertion_sort()
```
