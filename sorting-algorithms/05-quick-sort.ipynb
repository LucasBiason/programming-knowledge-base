# Quick Sort

Objetivo: Escolher um pivô, particionar a lista em elementos menores e maiores que o pivô, e ordenar recursivamente.

Complexidade:
- Tempo: O(n log n) caso médio, O(n²) pior caso
- Espaço: O(log n) para pilha de recursão

Categoria: Ordenação | Técnica: Divisão e Conquista

## Fundamentos

Quick sort é o "rei" dos algoritmos de ordenação na prática. Apesar de ter O(n²) no pior caso teórico, é extremamente rápido no caso médio e é o algoritmo padrão em muitas bibliotecas (C++ STL, Java para tipos primitivos).

A ideia central é o particionamento: escolher um elemento como pivô e reorganizar a lista de forma que elementos menores fiquem à esquerda e maiores à direita. O pivô fica em sua posição final. Depois, ordena recursivamente as duas partes.

Criado por Tony Hoare em 1959, revolucionou a computação. O nome "quick" vem do fato de ser, na prática, o algoritmo de ordenação por comparação mais rápido para a maioria dos casos reais.


```python
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

def quick_sort_animated(arr):
    """
    Gera animacao passo a passo do Quick Sort.
    Mostra escolha do pivo, particionamento e recursao.
    """
    arr = arr.copy()
    n = len(arr)
    frames = []

    frames.append({
        "arr": arr.copy(),
        "pivot": None,
        "comparing": [],
        "left_part": [],
        "right_part": [],
        "done": [],
        "label": "Array original"
    })

    done_positions = set()

    def quick_sort_track(arr, low, high):
        if low >= high:
            if 0 <= low < len(arr):
                done_positions.add(low)
            return

        pivot_val = arr[high]
        pivot_idx = high

        # Frame: pivo escolhido
        frames.append({
            "arr": arr.copy(),
            "pivot": pivot_idx,
            "comparing": [],
            "left_part": [],
            "right_part": [],
            "done": sorted(done_positions),
            "label": f"Pivo escolhido: [{pivot_idx}]={pivot_val}"
        })

        i = low
        for j in range(low, high):
            # Frame: comparando com pivo
            frames.append({
                "arr": arr.copy(),
                "pivot": pivot_idx,
                "comparing": [j],
                "left_part": list(range(low, i)),
                "right_part": list(range(i, j)),
                "done": sorted(done_positions),
                "label": f"Comparando [{j}]={arr[j]} com pivo {pivot_val}"
            })

            if arr[j] < pivot_val:
                if i != j:
                    arr[i], arr[j] = arr[j], arr[i]
                    frames.append({
                        "arr": arr.copy(),
                        "pivot": pivot_idx,
                        "comparing": [i, j],
                        "left_part": list(range(low, i + 1)),
                        "right_part": [],
                        "done": sorted(done_positions),
                        "label": f"Troca [{i}]={arr[i]} <-> [{j}]={arr[j]} (menor que pivo)"
                    })
                i += 1

        # Colocar pivo na posicao correta
        arr[i], arr[high] = arr[high], arr[i]
        done_positions.add(i)

        frames.append({
            "arr": arr.copy(),
            "pivot": i,
            "comparing": [],
            "left_part": list(range(low, i)),
            "right_part": list(range(i + 1, high + 1)),
            "done": sorted(done_positions),
            "label": f"Pivo {pivot_val} na posicao final [{i}]"
        })

        quick_sort_track(arr, low, i - 1)
        quick_sort_track(arr, i + 1, high)

    quick_sort_track(arr, 0, n - 1)

    # Frame final
    frames.append({
        "arr": arr.copy(),
        "pivot": None,
        "comparing": [],
        "left_part": [],
        "right_part": [],
        "done": list(range(n)),
        "label": "Vetor totalmente ordenado"
    })

    return frames


def animate_quick_sort(arr):
    """
    Renderiza a animacao do Quick Sort no Jupyter notebook.
    """
    frames = quick_sort_animated(arr)

    fig, ax = plt.subplots(figsize=(10, 4))
    max_val = max(arr) * 1.3

    def draw_frame(frame_idx):
        ax.clear()
        frame = frames[frame_idx]
        data = frame["arr"]
        n = len(data)
        colors = []

        for idx in range(n):
            if idx in frame["done"] and idx != frame.get("pivot"):
                colors.append("#10b981")    # verde - posicao final
            elif frame["pivot"] is not None and idx == frame["pivot"]:
                colors.append("#ef4444")    # vermelho - pivo
            elif idx in frame["comparing"]:
                colors.append("#f59e0b")    # amarelo - comparando
            elif idx in frame["left_part"]:
                colors.append("#8b5cf6")    # roxo - menores que pivo
            else:
                colors.append("#3b82f6")    # azul - normal

        bars = ax.bar(range(n), data, color=colors, edgecolor="black", linewidth=1.5)

        for idx, (bar, val) in enumerate(zip(bars, data)):
            ax.text(
                bar.get_x() + bar.get_width() / 2.0,
                bar.get_height() + max_val * 0.02,
                str(val),
                ha="center", va="bottom", fontsize=14, fontweight="bold"
            )

        ax.set_xlim(-0.5, n - 0.5)
        ax.set_ylim(0, max_val)
        ax.set_xticks(range(n))
        ax.set_xticklabels([f"[{i}]" for i in range(n)], fontsize=11)
        ax.set_ylabel("Valor", fontsize=11)
        ax.set_title(
            f"Quick Sort  |  Frame {frame_idx + 1}/{len(frames)}",
            fontsize=13, fontweight="bold", pad=12
        )

        ax.text(
            0.5, -0.12, frame["label"],
            transform=ax.transAxes, ha="center", fontsize=12,
            style="italic", color="#374151"
        )

        legend_elements = [
            mpatches.Patch(color="#3b82f6", label="Normal"),
            mpatches.Patch(color="#ef4444", label="Pivo"),
            mpatches.Patch(color="#f59e0b", label="Comparando"),
            mpatches.Patch(color="#8b5cf6", label="Menores que pivo"),
            mpatches.Patch(color="#10b981", label="Posicao final"),
        ]
        ax.legend(handles=legend_elements, loc="upper right", fontsize=9)
        ax.grid(axis="y", alpha=0.3, linestyle="--")
        fig.subplots_adjust(bottom=0.18)

    anim = FuncAnimation(
        fig, draw_frame,
        frames=len(frames),
        interval=800,
        repeat=True
    )
    plt.close(fig)
    return HTML(anim.to_jshtml())


# Executar animacao com o array de exemplo
animate_quick_sort([5, 3, 8, 4, 2])
```

## Como Funciona

Fase 1: Escolha do Pivô

1. Escolhe um elemento como pivô (geralmente o último)
2. Este elemento será colocado em sua posição final

Fase 2: Particionamento (Lomuto)

3. Mantém ponteiro `i` para elementos menores que o pivô
4. Ponteiro `j` percorre a lista
5. Se arr[j] < pivô: troca com arr[i] e avança i
6. No final, coloca pivô na posição i (posição final correta)

Fase 3: Recursão

7. Ordena recursivamente a parte esquerda (< pivô)
8. Ordena recursivamente a parte direita (> pivô)

Diferencial: Ordena in-place com particionamento inteligente. Não precisa de merge como o merge sort.


## Análise de Complexidade

### Tempo

Pior Caso: O(n²)
Ocorre quando o pivô é sempre o menor ou maior elemento (lista já ordenada com pivô no final). Cria partições desbalanceadas: n-1 elementos de um lado, 0 do outro.

Caso Médio: O(n log n) 
Com pivô razoavelmente balanceado, cria árvore de recursão com altura log n. Cada nível processa n elementos. Na prática, é o mais rápido!

Melhor Caso: O(n log n)
Quando o pivô sempre divide a lista exatamente ao meio. Comportamento idêntico ao merge sort.

### Espaço

O(log n)
Espaço para pilha de recursão. No pior caso (lista ordenada), pode ser O(n). Versões otimizadas garantem O(log n) ordenando primeiro a partição menor.

### Estabilidade

Não - o particionamento pode trocar elementos distantes, alterando a ordem relativa de elementos iguais.


```python
def quick_sort(arr, low=0, high=None):
    """
    Sorts an array using quick sort algorithm (Lomuto partition).
    
    Args:
        arr: List of comparable elements
        low: Starting index
        high: Ending index
        
    Returns:
        Sorted list (modified in-place)
        
    Time: O(n log n) average, O(n²) worst case
    Space: O(log n) for recursion stack
    """
    if high is None:
        high = len(arr) - 1
    
    # Base case: if partition has 0 or 1 element
    if low < high:
        # Partition and get pivot position
        pivot_idx = partition(arr, low, high)
        
        # Recursively sort left and right partitions
        quick_sort(arr, low, pivot_idx - 1)
        quick_sort(arr, pivot_idx + 1, high)
    
    return arr


def partition(arr, low, high):
    """
    Lomuto partition scheme.
    
    Args:
        arr: List to partition
        low: Starting index
        high: Ending index (pivot is arr[high])
        
    Returns:
        Final position of pivot
    """
    # Choose last element as pivot
    pivot = arr[high]
    
    # Pointer for elements smaller than pivot
    i = low - 1
    
    # Traverse and partition
    for j in range(low, high):
        if arr[j] < pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
    
    # Place pivot in correct position
    i += 1
    arr[i], arr[high] = arr[high], arr[i]
    
    return i
```

## Detalhes da Implementação

A função `partition` é o coração do algoritmo. Ela reorganiza a lista de forma que o pivô fique em sua posição final, com elementos menores à esquerda e maiores à direita.

O ponteiro `i` marca a fronteira entre elementos menores e maiores que o pivô. Quando encontramos um elemento menor, incrementamos `i` e trocamos.

A escolha do pivô afeta drasticamente a performance. Pivô aleatório ou mediana de três elementos evita o pior caso O(n²) em listas ordenadas.


```python
# Test with small array
test_arr = [10, 7, 8, 9, 1, 5]
print(f"Original: {test_arr}")

quick_sort(test_arr)
print(f"Sorted: {test_arr}")
```

```python
# Edge cases
empty = []
single = [42]
sorted_arr = [1, 2, 3, 4, 5]  # Worst case for Lomuto with last pivot
reverse = [5, 4, 3, 2, 1]
duplicates = [3, 1, 4, 1, 5, 9, 2, 6, 5]

test_cases = [
    ("Empty", empty),
    ("Single element", single),
    ("Already sorted", sorted_arr),
    ("Reverse sorted", reverse),
    ("With duplicates", duplicates)
]

for name, arr in test_cases:
    original = arr.copy()
    quick_sort(arr)
    print(f"{name}: {original} -> {arr}")
```

## Aplicações Práticas

### Quando Usar

- Caso geral - mais rápido na prática que merge sort
- Quando memória é limitada (O(log n) vs O(n) do merge sort)
- Listas grandes com dados aleatórios
- Implementações de bibliotecas padrão (C++ STL, Java primitivos)

### Quando Evitar

- Quando estabilidade é necessária
- Quando O(n log n) garantido é crucial (sistemas críticos)
- Listas já ordenadas (sem randomização do pivô)
- Sistemas com limite de profundidade de recursão


## Comparacao

| Algoritmo | Tempo (Pior) | Tempo (Melhor) | Espaco | Estavel | In-Place | Quando Usar |
|-----------|--------------|----------------|--------|---------|----------|-------------|
| Bubble Sort | O(n^2) | O(n) | O(1) | Sim | Sim | Educacional, listas muito pequenas |
| Selection Sort | O(n^2) | O(n^2) | O(1) | Nao | Sim | Quando swaps sao custosos |
| Insertion Sort | O(n^2) | O(n) | O(1) | Sim | Sim | Listas quase ordenadas, pequenas |
| Merge Sort | O(n log n) | O(n log n) | O(n) | Sim | Nao | Garantia de performance, estabilidade |
| Quick Sort | O(n^2) | O(n log n) | O(log n) | Nao | Sim | Caso geral rapido, melhor caso medio |

Quick sort vence no caso medio por ter constantes menores que merge sort e ser in-place. Merge sort vence quando precisamos de O(n log n) garantido ou estabilidade. Com pivo randomizado, quick sort raramente atinge O(n^2).


## Perguntas Comuns em Entrevistas

### Nível Básico

1. Qual a diferença entre quick sort e merge sort?
**R:** Quick sort particiona in-place e depois ordena recursivamente. Merge sort divide, ordena recursivamente e depois faz merge. Quick é mais rápido na prática mas tem pior caso O(n²).

2. O que é um pivô no quick sort?
**R:** Elemento escolhido para particionar a lista. Após o particionamento, o pivô está em sua posição final, com elementos menores à esquerda e maiores à direita.

### Nível Intermediário

1. Como evitar o pior caso O(n²) do quick sort?
**R:** Escolher pivô aleatório ou usar mediana de três elementos (primeiro, meio, último). Isso torna extremamente improvável ter partições desbalanceadas consistentemente.

### Nível Avançado

1. Por que quick sort é mais rápido que merge sort na prática se ambos são O(n log n)?
**R:** Quick sort tem constantes menores (menos operações por comparação), melhor localidade de cache (in-place), e não precisa alocar memória extra. O overhead do merge sort (criar arrays temporários) é significativo.

## Exercicios

**Desafio 1:** Implemente o Quick Sort com pivô aleatorio (randomized pivot) para evitar o pior caso O(n^2) em listas ja ordenadas.

**Desafio 2:** Implemente a particao de tres vias (Dutch National Flag) que trata elementos iguais ao pivo separadamente, otimizando arrays com muitos duplicados.

```python
import random

# Desafio 1: Quick Sort com pivo aleatorio
def quick_sort_random(arr, low=0, high=None):
    if high is None:
        high = len(arr) - 1
    if low < high:
        # Troca pivo aleatorio com o ultimo elemento
        rand_idx = random.randint(low, high)
        arr[rand_idx], arr[high] = arr[high], arr[rand_idx]
        pivot_idx = partition_lomuto(arr, low, high)
        quick_sort_random(arr, low, pivot_idx - 1)
        quick_sort_random(arr, pivot_idx + 1, high)
    return arr

def partition_lomuto(arr, low, high):
    pivot = arr[high]
    i = low - 1
    for j in range(low, high):
        if arr[j] < pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    return i + 1

# Desafio 2: Particao de tres vias (Dutch National Flag)
def quick_sort_3way(arr, low=0, high=None):
    if high is None:
        high = len(arr) - 1
    if low >= high:
        return arr
    pivot = arr[low + (high - low) // 2]
    lt = low   # arr[low..lt-1] < pivot
    gt = high  # arr[gt+1..high] > pivot
    i = low
    while i <= gt:
        if arr[i] < pivot:
            arr[lt], arr[i] = arr[i], arr[lt]
            lt += 1; i += 1
        elif arr[i] > pivot:
            arr[i], arr[gt] = arr[gt], arr[i]
            gt -= 1
        else:
            i += 1
    # Elementos iguais ao pivo ja estao no lugar
    quick_sort_3way(arr, low, lt - 1)
    quick_sort_3way(arr, gt + 1, high)
    return arr

# Teste
arr1 = [10, 7, 8, 9, 1, 5]
print("Quick Sort (pivo aleatorio):", quick_sort_random(arr1.copy()))

arr2 = [3, 3, 1, 2, 3, 1, 2, 3, 3]
print("Quick Sort 3 vias (muitos duplicados):", quick_sort_3way(arr2.copy()))
```

## Benchmark

Comparacao de desempenho com o `sorted()` nativo do Python para arrays de tamanhos 100, 1.000 e 5.000 elementos.

```python
import timeit
import random

def benchmark_quick_sort():
    sizes = [100, 1000, 5000]
    print(f"{'Tamanho':>10} | {'quick_sort (ms)':>17} | {'sorted() (ms)':>15}")
    print("-" * 50)
    for n in sizes:
        data = random.sample(range(n * 10), n)

        t_quick = timeit.timeit(
            lambda d=data: quick_sort(d.copy()), number=5
        ) / 5 * 1000

        t_sorted = timeit.timeit(
            lambda d=data: sorted(d), number=5
        ) / 5 * 1000

        print(f"{n:>10} | {t_quick:>17.4f} | {t_sorted:>15.4f}")

benchmark_quick_sort()
```
