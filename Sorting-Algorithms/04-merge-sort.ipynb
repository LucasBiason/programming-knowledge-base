# ğŸ”€ Merge Sort

Divide a lista em sublistas menores, ordena recursivamente e combina (merge) as partes ordenadas.

**Complexidade:**
- â±ï¸ Tempo: O(n log n) em todos os casos
- ğŸ’¾ EspaÃ§o: O(n)

**Categoria:** OrdenaÃ§Ã£o | **TÃ©cnica:** DivisÃ£o e Conquista


## ğŸ“š Fundamentos

Merge sort Ã© o primeiro algoritmo "profissional" da lista. Enquanto bubble, selection e insertion sÃ£o O(nÂ²), o merge sort alcanÃ§a O(n log n) usando a tÃ©cnica de **DivisÃ£o e Conquista**.

A analogia perfeita Ã© corrigir provas ğŸ“: em vez de corrigir 100 provas de uma vez, vocÃª divide em 2 pilhas de 50, depois em 4 pilhas de 25, atÃ© ter pilhas de 1 prova. Depois combina as pilhas jÃ¡ corrigidas de forma ordenada.

Criado por John von Neumann em 1945, Ã© um dos algoritmos mais importantes da computaÃ§Ã£o. Ã‰ estÃ¡vel, previsÃ­vel (sempre O(n log n)) e forma a base do Timsort usado no Python.

## âš™ï¸ Como Funciona

### ğŸ”ª Fase 1: DivisÃ£o (Divide)

1. âœ‚ï¸ Divide a lista ao meio recursivamente
2. ğŸ” Continua dividindo atÃ© ter sublistas de 1 elemento
3. ğŸ“¦ Sublistas de 1 elemento estÃ£o automaticamente ordenadas

### ğŸ”— Fase 2: Conquista (Merge)

4. ğŸ”€ Combina duas sublistas ordenadas em uma lista maior ordenada
5. ğŸ” Compara os primeiros elementos de cada sublista
6. ğŸ“¥ Insere o menor no resultado
7. ğŸ” Repete atÃ© processar todos os elementos

ğŸ’¡ **Chave:** A fase de merge Ã© O(n) e ocorre em O(log n) nÃ­veis da Ã¡rvore de recursÃ£o, resultando em O(n log n) total.

## ğŸ“ˆ AnÃ¡lise de Complexidade

### â±ï¸ Tempo

**ğŸŸ¢ Pior Caso: O(n log n) â­**
Sempre divide a lista ao meio (log n nÃ­veis) e faz merge de n elementos em cada nÃ­vel. NÃ£o importa a entrada, a complexidade Ã© sempre a mesma.

**ğŸŸ¡ Caso MÃ©dio: O(n log n)**
Comportamento idÃªntico ao pior caso.

**ğŸŸ¢ Melhor Caso: O(n log n)**
Mesmo para listas jÃ¡ ordenadas, o algoritmo divide e faz merge. NÃ£o hÃ¡ otimizaÃ§Ã£o para casos melhores.

### ğŸ’¾ EspaÃ§o

**O(n)**
Requer array auxiliar temporÃ¡rio para o merge. Esse Ã© o "preÃ§o" pago pela garantia de O(n log n).

### âš–ï¸ Estabilidade

âœ… Sim - durante o merge, quando elementos sÃ£o iguais, pegamos primeiro da sublista esquerda, mantendo a ordem original.


```python
def merge_sort(arr):
    """
    Sorts an array using merge sort algorithm.
    
    Args:
        arr: List of comparable elements
        
    Returns:
        Sorted list (new list created)
        
    Time: O(n log n) in all cases
    Space: O(n)
    """
    # Base case: list with 0 or 1 element is already sorted
    if len(arr) <= 1:
        return arr
    
    # Divide: split list in half
    mid = len(arr) // 2
    left = arr[:mid]
    right = arr[mid:]
    
    # Conquer: recursively sort both halves
    left = merge_sort(left)
    right = merge_sort(right)
    
    # Combine: merge sorted halves
    return merge(left, right)


def merge(left, right):
    """
    Merges two sorted lists into one sorted list.
    
    Args:
        left: Sorted list
        right: Sorted list
        
    Returns:
        Merged sorted list
    """
    result = []
    i = j = 0
    
    # Compare elements from both lists and add smaller to result
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:  # <= ensures stability
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    
    # Add remaining elements (one list will be empty)
    result.extend(left[i:])
    result.extend(right[j:])
    
    return result
```

## ğŸ’¡ Detalhes da ImplementaÃ§Ã£o

ğŸ”ª A funÃ§Ã£o `merge_sort` divide recursivamente atÃ© ter sublistas de 1 elemento (caso base). A recursÃ£o cria uma Ã¡rvore de chamadas com altura log n.

ğŸ”— A funÃ§Ã£o `merge` Ã© onde acontece a "mÃ¡gica". Ela combina duas listas ordenadas em uma lista maior ordenada, comparando sempre os primeiros elementos de cada lista.

âš¡ A condiÃ§Ã£o `left[i] <= right[j]` (e nÃ£o `<`) garante estabilidade: quando elementos sÃ£o iguais, pegamos primeiro da lista esquerda, preservando a ordem original.


```python
# Test with small array
test_arr = [38, 27, 43, 3, 9, 82, 10]
print(f"Original: {test_arr}")

sorted_arr = merge_sort(test_arr)
print(f"Sorted: {sorted_arr}")
```

```python
# Edge cases
empty = []
single = [42]
sorted_arr = [1, 2, 3, 4, 5]
reverse = [5, 4, 3, 2, 1]
duplicates = [3, 1, 4, 1, 5, 9, 2, 6, 5]

test_cases = [
    ("Empty", empty),
    ("Single element", single),
    ("Already sorted", sorted_arr),
    ("Reverse sorted", reverse),
    ("With duplicates", duplicates)
]

for name, arr in test_cases:
    sorted_result = merge_sort(arr)
    print(f"{name}: {arr} -> {sorted_result}")
```

## ğŸ’¼ Perguntas Comuns em Entrevistas

### ğŸŸ¢ NÃ­vel BÃ¡sico

1. O que Ã© "DivisÃ£o e Conquista"?
   **R:** TÃ©cnica que divide o problema em subproblemas menores, resolve cada um recursivamente e combina as soluÃ§Ãµes. No merge sort: divide em metades, ordena recursivamente, faz merge.

2. Por que merge sort Ã© O(n log n)?
   **R:** A Ã¡rvore de recursÃ£o tem altura log n (dividindo ao meio) e cada nÃ­vel processa n elementos no merge. Total: n Ã— log n = O(n log n).

### ğŸŸ¡ NÃ­vel IntermediÃ¡rio

1. Como implementar merge sort in-place para economizar memÃ³ria?
   **R:** Ã‰ muito complexo e perde a vantagem de simplicidade. Existem versÃµes in-place mas sÃ£o raramente usadas. Na prÃ¡tica, se memÃ³ria Ã© crÃ­tica, usa-se heap sort (O(n log n) com O(1) espaÃ§o).

### ğŸ”´ NÃ­vel AvanÃ§ado

1. Como adaptar merge sort para ordenaÃ§Ã£o externa (dados em disco)?
   **R:** External merge sort: divide dados em chunks que cabem na memÃ³ria, ordena cada chunk, depois faz k-way merge dos chunks ordenados lendo do disco. Usado em bancos de dados para ordenar tabelas gigantes.
