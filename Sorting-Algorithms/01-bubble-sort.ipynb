# ğŸ”„ Bubble Sort

Ordena uma lista comparando elementos adjacentes e trocando-os quando estÃ£o fora de ordem.

**Complexidade:**
- â±ï¸ Tempo: O(nÂ²) no pior caso, O(n) no melhor caso
- ğŸ’¾ EspaÃ§o: O(1)

## ğŸ“š Fundamentos

Bubble sort Ã© o algoritmo de ordenaÃ§Ã£o mais simples de entender. Compara elementos adjacentes e troca-os se estiverem na ordem errada. O processo repete atÃ© nÃ£o haver mais trocas necessÃ¡rias.

O nome vem do comportamento: os maiores elementos "sobem" para o final da lista, como bolhas ğŸ«§ subindo na Ã¡gua. A cada passagem completa, pelo menos um elemento vai para sua posiÃ§Ã£o final correta.

Ã‰ um dos primeiros algoritmos ensinados em cursos de ciÃªncia da computaÃ§Ã£o por sua simplicidade, apesar da baixa eficiÃªncia em listas grandes.


```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np

def visualize_bubble_sort_step(arr, step_num, comparing_idx=None, swapped=False):
    """
    Visualiza um passo do bubble sort de forma simples e clara.
    """
    fig, ax = plt.subplots(figsize=(10, 3))
    
    n = len(arr)
    colors = ['#3b82f6'] * n
    
    # Destacar elementos sendo comparados
    if comparing_idx is not None and comparing_idx < n - 1:
        colors[comparing_idx] = '#fb923c'  # laranja - comparando
        colors[comparing_idx + 1] = '#fb923c'  # laranja - comparando
    
    # Destacar elementos jÃ¡ ordenados (verde)
    # Na Ãºltima posiÃ§Ã£o apÃ³s cada passagem completa
    if step_num > 0:
        colors[-step_num] = '#10b981'  # verde - jÃ¡ ordenado
    
    # Criar barras
    bars = ax.bar(range(n), arr, color=colors, edgecolor='black', linewidth=1.5)
    
    # Adicionar valores nas barras
    for i, (bar, val) in enumerate(zip(bars, arr)):
        height = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2., height,
                f'{val}', ha='center', va='bottom', fontsize=12, fontweight='bold')
    
    # ConfiguraÃ§Ãµes do grÃ¡fico
    ax.set_xlim(-0.5, n - 0.5)
    ax.set_ylim(0, max(arr) * 1.2)
    ax.set_xticks(range(n))
    ax.set_xticklabels([f'[{i}]' for i in range(n)], fontsize=10)
    ax.set_ylabel('Valor', fontsize=11)
    ax.set_title(f'Bubble Sort - Passo {step_num}: Array [5, 3, 8, 4, 2]', 
                 fontsize=13, fontweight='bold', pad=15)
    ax.grid(axis='y', alpha=0.3, linestyle='--')
    
    # Legenda
    legend_elements = [
        mpatches.Patch(color='#3b82f6', label='Normal'),
        mpatches.Patch(color='#fb923c', label='Comparando'),
        mpatches.Patch(color='#10b981', label='Ordenado')
    ]
    ax.legend(handles=legend_elements, loc='upper right', fontsize=9)
    
    plt.tight_layout()
    plt.show()

# Visualizar primeira passagem completa
arr = [5, 3, 8, 4, 2]
print("=== PRIMEIRA PASSAGEM COMPLETA ===\n")

# Passo 1: Compara [0] e [1]
visualize_bubble_sort_step([5, 3, 8, 4, 2], 1, comparing_idx=0)
print("Comparando 5 e 3 â†’ 5 > 3, troca â†’ [3, 5, 8, 4, 2]\n")

# Passo 2: Compara [1] e [2]
visualize_bubble_sort_step([3, 5, 8, 4, 2], 2, comparing_idx=1)
print("Comparando 5 e 8 â†’ 5 < 8, mantÃ©m â†’ [3, 5, 8, 4, 2]\n")

# Passo 3: Compara [2] e [3]
visualize_bubble_sort_step([3, 5, 8, 4, 2], 3, comparing_idx=2)
print("Comparando 8 e 4 â†’ 8 > 4, troca â†’ [3, 5, 4, 8, 2]\n")

# Passo 4: Compara [3] e [4]
visualize_bubble_sort_step([3, 5, 4, 8, 2], 4, comparing_idx=3)
print("Comparando 8 e 2 â†’ 8 > 2, troca â†’ [3, 5, 4, 2, 8]\n")
print("âœ… Maior elemento (8) 'borbulhou' para o final!")
```

## âš™ï¸ Como Funciona

1. â¡ï¸ Percorre a lista do inÃ­cio ao fim
2. ğŸ” Compara cada elemento com seu vizinho da direita
3. ğŸ”„ Se o elemento da esquerda for maior, troca-os
4. ğŸ” Repete o processo atÃ© nÃ£o haver mais trocas

ğŸ’¡ **OtimizaÃ§Ã£o:** usar uma flag para detectar se houve trocas na passagem atual. Se nÃ£o houve trocas, a lista jÃ¡ estÃ¡ ordenada e podemos parar.

## ğŸ“ˆ AnÃ¡lise de Complexidade

### â±ï¸ Tempo

**ğŸ”´ Pior Caso: O(nÂ²)**
Ocorre quando a lista estÃ¡ inversamente ordenada. Cada elemento precisa ser movido atÃ© sua posiÃ§Ã£o final, resultando em n passagens com n comparaÃ§Ãµes cada.

**ğŸŸ¡ Caso MÃ©dio: O(nÂ²)**
Mesmo comportamento do pior caso na maioria dos cenÃ¡rios.

**ğŸŸ¢ Melhor Caso: O(n)**
Quando a lista jÃ¡ estÃ¡ ordenada. Com a otimizaÃ§Ã£o da flag, apenas uma passagem Ã© necessÃ¡ria para confirmar que nÃ£o hÃ¡ trocas.

### ğŸ’¾ EspaÃ§o

**O(1)**
O algoritmo ordena in-place, usando apenas variÃ¡veis auxiliares para trocas. NÃ£o requer espaÃ§o adicional proporcional ao tamanho da entrada.

### âš–ï¸ Estabilidade

âœ… Sim - elementos com valores iguais mantÃªm sua ordem relativa original.


```python
def bubble_sort(arr):
    """
    Sorts an array using bubble sort algorithm.
    
    Args:
        arr: List of comparable elements
        
    Returns:
        Sorted list (modified in-place)
        
    Time: O(nÂ²) worst case, O(n) best case
    Space: O(1)
    """
    n = len(arr)
    
    # Outer loop: number of passes needed
    for i in range(n):
        swapped = False
        
        # Inner loop: compare adjacent elements
        # After each pass, largest element is in place
        # So we can reduce the range by i each time
        for j in range(0, n - 1 - i):
            if arr[j] > arr[j + 1]:
                # Swap elements
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        
        # Optimization: if no swaps, list is sorted
        if not swapped:
            break
    
    return arr
```

## ğŸ’¡ Detalhes da ImplementaÃ§Ã£o

ğŸ”„ O loop externo controla quantas passagens completas fazemos pela lista. O loop interno compara pares adjacentes e troca quando necessÃ¡rio.

âš¡ A otimizaÃ§Ã£o com a flag `swapped` permite terminar antecipadamente quando a lista jÃ¡ estÃ¡ ordenada. Sem essa otimizaÃ§Ã£o, o algoritmo sempre faz n passagens completas, mesmo quando desnecessÃ¡rio.

ğŸ¯ A reduÃ§Ã£o do range interno (`n - 1 - i`) Ã© uma segunda otimizaÃ§Ã£o. Como a cada passagem o maior elemento vai para o final, nÃ£o precisamos comparar esses elementos jÃ¡ ordenados nas prÃ³ximas iteraÃ§Ãµes.


```python
# Test with small array
test_arr = [5, 3, 8, 4, 2]
print(f"Original: {test_arr}")

bubble_sort(test_arr)
print(f"Sorted: {test_arr}")
```

```python
# Edge cases
empty = []
single = [42]
sorted_arr = [1, 2, 3, 4, 5]
reverse = [5, 4, 3, 2, 1]
duplicates = [3, 1, 4, 1, 5, 9, 2, 6, 5]

test_cases = [
    ("Empty", empty),
    ("Single element", single),
    ("Already sorted", sorted_arr),
    ("Reverse sorted", reverse),
    ("With duplicates", duplicates)
]

for name, arr in test_cases:
    original = arr.copy()
    bubble_sort(arr)
    print(f"{name}: {original} -> {arr}")
```

## âš–ï¸ ComparaÃ§Ã£o

| Algoritmo | Tempo (Pior) | EspaÃ§o | EstÃ¡vel | Quando Usar |
|-----------|--------------|--------|---------|-------------|
| ğŸ«§ Bubble Sort | O(nÂ²) | O(1) | âœ… | Listas muito pequenas |

ğŸ’¡ Bubble sort Ã© o mais lento entre os algoritmos bÃ¡sicos. Selection sort tem a mesma complexidade mas faz menos trocas. Insertion sort Ã© melhor para listas quase ordenadas. Para listas grandes, algoritmos O(n log n) como merge sort e quick sort sÃ£o necessÃ¡rios.
