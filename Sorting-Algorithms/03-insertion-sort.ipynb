## ğŸ“š Fundamentos

Insertion sort Ã© o algoritmo mais "intuitivo" dos trÃªs bÃ¡sicos. Ã‰ exatamente como a maioria das pessoas organiza cartas de baralho na mÃ£o: pega uma carta nova e insere na posiÃ§Ã£o correta entre as cartas jÃ¡ ordenadas.

A analogia do baralho ğŸƒ Ã© perfeita: sua mÃ£o esquerda segura cartas jÃ¡ ordenadas, sua mÃ£o direita pega uma nova carta da mesa, vocÃª desliza as cartas maiores para a direita e insere a nova carta no espaÃ§o criado.

Ã‰ o algoritmo usado por algoritmos industriais como Timsort (Python/Java) para ordenar sublistas pequenas, pois para listas pequenas ou quase ordenadas, ele supera atÃ© o quick sort devido Ã  baixa sobrecarga.


```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np

def visualize_insertion_sort_step(arr, key_idx, comparing_idx=None, shifting=False):
    """
    Visualiza um passo do insertion sort de forma simples e clara.
    """
    fig, ax = plt.subplots(figsize=(10, 3))
    
    n = len(arr)
    colors = ['#10b981'] * n  # verde - parte ordenada
    
    # Destacar chave sendo inserida (amarelo)
    colors[key_idx] = '#fbbf24'  # amarelo - chave
    
    # Destacar elementos sendo deslocados (azul)
    if shifting and comparing_idx is not None:
        for i in range(comparing_idx + 1, key_idx):
            colors[i] = '#3b82f6'  # azul - deslocando
    
    # Destacar elemento sendo comparado (laranja)
    if comparing_idx is not None and comparing_idx < key_idx:
        colors[comparing_idx] = '#fb923c'  # laranja - comparando
    
    # Criar barras
    bars = ax.bar(range(n), arr, color=colors, edgecolor='black', linewidth=1.5)
    
    # Adicionar valores nas barras
    for i, (bar, val) in enumerate(zip(bars, arr)):
        height = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2., height,
                f'{val}', ha='center', va='bottom', fontsize=12, fontweight='bold')
    
    # Seta indicando deslocamento
    if shifting and comparing_idx is not None:
        for i in range(comparing_idx + 1, key_idx + 1):
            ax.annotate('', xy=(i, arr[i] + max(arr) * 0.1), 
                       xytext=(i-1, arr[i] + max(arr) * 0.1),
                       arrowprops=dict(arrowstyle='->', color='#3b82f6', lw=2))
    
    # ConfiguraÃ§Ãµes do grÃ¡fico
    ax.set_xlim(-0.5, n - 0.5)
    ax.set_ylim(0, max(arr) * 1.3)
    ax.set_xticks(range(n))
    ax.set_xticklabels([f'[{i}]' for i in range(n)], fontsize=10)
    ax.set_ylabel('Valor', fontsize=11)
    ax.set_title(f'Insertion Sort - Inserindo elemento {arr[key_idx]} (Ã­ndice {key_idx})', 
                 fontsize=13, fontweight='bold', pad=15)
    ax.grid(axis='y', alpha=0.3, linestyle='--')
    
    # Linha divisÃ³ria
    if key_idx > 0:
        ax.axvline(x=key_idx - 0.5, color='gray', linestyle='--', linewidth=2, alpha=0.5)
        ax.text(key_idx - 0.5, max(arr) * 1.2, 'Ordenado', 
                ha='center', fontsize=9, style='italic')
    
    # Legenda
    legend_elements = [
        mpatches.Patch(color='#10b981', label='Ordenado'),
        mpatches.Patch(color='#fbbf24', label='Chave (inserindo)'),
        mpatches.Patch(color='#3b82f6', label='Deslocando'),
        mpatches.Patch(color='#fb923c', label='Comparando')
    ]
    ax.legend(handles=legend_elements, loc='upper right', fontsize=9)
    
    plt.tight_layout()
    plt.show()

# Visualizar inserÃ§Ãµes
arr = [5, 2, 4, 6, 1]
print("=== INSERÃ‡Ã•ES DO INSERTION SORT ===\n")

# Inserir 2 (Ã­ndice 1)
visualize_insertion_sort_step([5, 2, 4, 6, 1], 1, comparing_idx=0, shifting=True)
print("Inserindo 2: Compara com 5 â†’ 2 < 5, desloca 5 â†’ [2, 5, 4, 6, 1]\n")

# Inserir 4 (Ã­ndice 2)
visualize_insertion_sort_step([2, 5, 4, 6, 1], 2, comparing_idx=1, shifting=True)
print("Inserindo 4: Compara com 5 â†’ 4 < 5, desloca 5 â†’ [2, 4, 5, 6, 1]\n")

# Inserir 6 (Ã­ndice 3) - jÃ¡ estÃ¡ no lugar
visualize_insertion_sort_step([2, 4, 5, 6, 1], 3, comparing_idx=2)
print("Inserindo 6: Compara com 5 â†’ 6 > 5, jÃ¡ no lugar â†’ [2, 4, 5, 6, 1]\n")

# Inserir 1 (Ã­ndice 4)
visualize_insertion_sort_step([2, 4, 5, 6, 1], 4, comparing_idx=3, shifting=True)
print("Inserindo 1: Compara com 6,5,4,2 â†’ desloca todos â†’ [1, 2, 4, 5, 6]\n")
print("âœ… Array ordenado!")
```

## âš™ï¸ Como Funciona

1. âœ‹ Assume que o primeiro elemento jÃ¡ estÃ¡ "ordenado" (lista de 1 elemento)
2. ğŸ†• Pega o prÃ³ximo elemento (chave) da parte nÃ£o ordenada
3. ğŸ‘ˆ Compara com elementos Ã  esquerda, da direita para esquerda
4. â¡ï¸ Desloca elementos maiores uma posiÃ§Ã£o Ã  direita
5. ğŸ“ Insere a chave no espaÃ§o criado
6. ğŸ” Repete atÃ© processar todos os elementos

ğŸ’¡ **Diferencial:** Usa deslocamento (shift) em vez de trocas (swap). O loop `while` desloca elementos maiores atÃ© encontrar a posiÃ§Ã£o correta.

## ğŸ“ˆ AnÃ¡lise de Complexidade

### â±ï¸ Tempo

**ğŸ”´ Pior Caso: O(nÂ²)**
Ocorre quando a lista estÃ¡ inversamente ordenada. Cada elemento precisa ser comparado e deslocado atravÃ©s de toda a parte ordenada.

**ğŸŸ¡ Caso MÃ©dio: O(nÂ²)**
Em mÃ©dia, cada elemento precisa ser comparado com metade dos elementos jÃ¡ ordenados.

**ğŸŸ¢ Melhor Caso: O(n) â­**
Quando a lista jÃ¡ estÃ¡ ordenada ou quase ordenada. O loop `while` nunca executa, fazendo apenas uma comparaÃ§Ã£o por elemento. Por isso Ã© usado no Timsort!

### ğŸ’¾ EspaÃ§o

**O(1)**
Ordena in-place, usando apenas variÃ¡veis auxiliares (chave e Ã­ndice). NÃ£o requer espaÃ§o adicional.

### âš–ï¸ Estabilidade

âœ… Sim - elementos iguais mantÃªm sua ordem relativa porque sÃ³ deslocamos quando o elemento Ã© estritamente menor.


```python
def insertion_sort(arr):
    """
    Sorts an array using insertion sort algorithm.
    
    Args:
        arr: List of comparable elements
        
    Returns:
        Sorted list (modified in-place)
        
    Time: O(nÂ²) worst case, O(n) best case
    Space: O(1)
    """
    n = len(arr)
    
    # Start from second element (index 1)
    # First element is considered already sorted
    for i in range(1, n):
        # Save the current element (key)
        key = arr[i]
        
        # Start comparing with element to the left
        j = i - 1
        
        # Shift loop: move elements greater than key to the right
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]  # Shift right
            j -= 1
        
        # Insert key in the correct position
        arr[j + 1] = key
    
    return arr
```

## ğŸ’¡ Detalhes da ImplementaÃ§Ã£o

ğŸ”„ O loop externo comeÃ§a do segundo elemento (Ã­ndice 1), pois assumimos que o primeiro elemento sozinho jÃ¡ estÃ¡ "ordenado".

ğŸ¯ A variÃ¡vel `key` guarda o elemento que queremos inserir. O loop `while` desloca elementos maiores uma posiÃ§Ã£o Ã  direita, criando um "espaÃ§o" para a chave.

âš¡ A condiÃ§Ã£o `key < arr[j]` (e nÃ£o `<=`) garante estabilidade: elementos iguais nÃ£o sÃ£o deslocados, mantendo a ordem original.


```python
# Test with small array
test_arr = [5, 2, 4, 6, 1]
print(f"Original: {test_arr}")

insertion_sort(test_arr)
print(f"Sorted: {test_arr}")
```

```python
# Edge cases
empty = []
single = [42]
sorted_arr = [1, 2, 3, 4, 5]
reverse = [5, 4, 3, 2, 1]
duplicates = [3, 1, 4, 1, 5, 9, 2, 6, 5]
almost_sorted = [1, 2, 3, 5, 4]  # Best case scenario

test_cases = [
    ("Empty", empty),
    ("Single element", single),
    ("Already sorted", sorted_arr),
    ("Reverse sorted", reverse),
    ("With duplicates", duplicates),
    ("Almost sorted", almost_sorted)
]

for name, arr in test_cases:
    original = arr.copy()
    insertion_sort(arr)
    print(f"{name}: {original} -> {arr}")
```

## ğŸ¯ AplicaÃ§Ãµes PrÃ¡ticas

### âœ… Quando Usar

- ğŸš€ Listas pequenas (menos de 20 elementos) - mais rÃ¡pido que quick/merge sort

### âŒ Quando Evitar

- âš¡ AplicaÃ§Ãµes que precisam de O(n log n) garantido

ğŸ’¡ **Destaque:** Insertion sort tem o melhor caso O(n), tornando-o ideal para dados quase ordenados. Ã‰ por isso que o Timsort usa insertion sort para sublistas pequenas - na prÃ¡tica, muitos dados reais tÃªm "runs" parcialmente ordenados.
