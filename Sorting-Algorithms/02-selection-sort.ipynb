# ğŸ¯ Selection Sort

**Complexidade:**
- â±ï¸ Tempo: O(nÂ²) em todos os casos
- ğŸ’¾ EspaÃ§o: O(1)

## ğŸ“š Fundamentos

Selection sort Ã© o algoritmo que "seleciona" o menor elemento em cada iteraÃ§Ã£o e coloca-o na posiÃ§Ã£o correta. Diferente do bubble sort que faz mÃºltiplas trocas, o selection sort faz apenas uma troca por passagem.

A analogia perfeita Ã© selecionar jogadores para um time: vocÃª varre todos os candidatos, encontra o melhor, e coloca-o na primeira posiÃ§Ã£o. Depois repete para a segunda posiÃ§Ã£o, terceira, e assim por diante.

Ã‰ interessante quando o custo de escrita na memÃ³ria Ã© alto (swaps custam caro), pois faz exatamente n-1 trocas no total - o mÃ­nimo possÃ­vel para qualquer algoritmo de ordenaÃ§Ã£o por comparaÃ§Ã£o.


```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np

def visualize_selection_sort_step(arr, iteration, min_idx=None, current_pos=None):
    """
    Visualiza um passo do selection sort de forma simples e clara.
    """
    fig, ax = plt.subplots(figsize=(10, 3))
    
    n = len(arr)
    colors = ['#f97316'] * n
    
    # Destacar parte ordenada (verde)
    if iteration > 0:
        for i in range(iteration):
            colors[i] = '#10b981'  # verde - jÃ¡ ordenado
    
    # Destacar menor elemento encontrado (vermelho)
    if min_idx is not None:
        colors[min_idx] = '#ef4444'  # vermelho - menor elemento
    
    # Destacar posiÃ§Ã£o atual sendo preenchida (amarelo)
    if current_pos is not None:
        if colors[current_pos] != '#ef4444':  # nÃ£o sobrescrever se for o mÃ­nimo
            colors[current_pos] = '#fbbf24'  # amarelo - posiÃ§Ã£o atual
    
    # Criar barras
    bars = ax.bar(range(n), arr, color=colors, edgecolor='black', linewidth=1.5)
    
    # Adicionar valores nas barras
    for i, (bar, val) in enumerate(zip(bars, arr)):
        height = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2., height,
                f'{val}', ha='center', va='bottom', fontsize=12, fontweight='bold')
    
    # ConfiguraÃ§Ãµes do grÃ¡fico
    ax.set_xlim(-0.5, n - 0.5)
    ax.set_ylim(0, max(arr) * 1.2)
    ax.set_xticks(range(n))
    ax.set_xticklabels([f'[{i}]' for i in range(n)], fontsize=10)
    ax.set_ylabel('Valor', fontsize=11)
    ax.set_title(f'Selection Sort - IteraÃ§Ã£o {iteration + 1}: Array [29, 10, 14, 37, 13]', 
                 fontsize=13, fontweight='bold', pad=15)
    ax.grid(axis='y', alpha=0.3, linestyle='--')
    
    # Linha divisÃ³ria entre ordenado e nÃ£o ordenado
    if iteration > 0:
        ax.axvline(x=iteration - 0.5, color='gray', linestyle='--', linewidth=2, alpha=0.5)
        ax.text(iteration - 0.5, max(arr) * 1.1, 'Ordenado', 
                ha='center', fontsize=9, style='italic')
    
    # Legenda
    legend_elements = [
        mpatches.Patch(color='#f97316', label='NÃ£o ordenado'),
        mpatches.Patch(color='#10b981', label='Ordenado'),
        mpatches.Patch(color='#ef4444', label='Menor elemento'),
        mpatches.Patch(color='#fbbf24', label='PosiÃ§Ã£o atual')
    ]
    ax.legend(handles=legend_elements, loc='upper right', fontsize=9)
    
    plt.tight_layout()
    plt.show()

# Visualizar iteraÃ§Ãµes
arr = [29, 10, 14, 37, 13]
print("=== ITERAÃ‡Ã•ES DO SELECTION SORT ===\n")

# IteraÃ§Ã£o 1: Encontra menor (10) e troca com posiÃ§Ã£o 0
visualize_selection_sort_step([29, 10, 14, 37, 13], 0, min_idx=1, current_pos=0)
print("Menor elemento: 10 (Ã­ndice 1)")
print("Troca 29 â†” 10 â†’ [10, 29, 14, 37, 13]\n")

# IteraÃ§Ã£o 2: Encontra menor (13) e troca com posiÃ§Ã£o 1
visualize_selection_sort_step([10, 29, 14, 37, 13], 1, min_idx=4, current_pos=1)
print("Menor elemento: 13 (Ã­ndice 4)")
print("Troca 29 â†” 13 â†’ [10, 13, 14, 37, 29]\n")

# IteraÃ§Ã£o 3: Encontra menor (14) jÃ¡ estÃ¡ na posiÃ§Ã£o correta
visualize_selection_sort_step([10, 13, 14, 37, 29], 2, min_idx=2, current_pos=2)
print("Menor elemento: 14 (Ã­ndice 2) - jÃ¡ na posiÃ§Ã£o correta\n")

# IteraÃ§Ã£o 4: Encontra menor (29) e troca com posiÃ§Ã£o 3
visualize_selection_sort_step([10, 13, 14, 37, 29], 3, min_idx=4, current_pos=3)
print("Menor elemento: 29 (Ã­ndice 4)")
print("Troca 37 â†” 29 â†’ [10, 13, 14, 29, 37]\n")
print("âœ… Array ordenado!")
```

## âš™ï¸ Como Funciona

1. ğŸ” Encontra o menor elemento na parte nÃ£o ordenada
2. ğŸ”„ Troca com o primeiro elemento da parte nÃ£o ordenada
3. â¡ï¸ Move a fronteira entre ordenado/nÃ£o-ordenado uma posiÃ§Ã£o Ã  direita
4. ğŸ” Repete atÃ© toda a lista estar ordenada

ğŸ’¡ **CaracterÃ­stica:** Faz exatamente n-1 trocas, independente da entrada. Mesmo que a lista jÃ¡ esteja ordenada, o algoritmo continua procurando o mÃ­nimo.

## ğŸ“ˆ AnÃ¡lise de Complexidade

### â±ï¸ Tempo

**ğŸ”´ Pior Caso: O(nÂ²)**
Mesmo que a lista esteja ordenada, o algoritmo varre toda a parte nÃ£o ordenada procurando o mÃ­nimo. Sempre faz (n-1) + (n-2) + ... + 1 = n(n-1)/2 comparaÃ§Ãµes.

**ğŸŸ¡ Caso MÃ©dio: O(nÂ²)**
Comportamento idÃªntico ao pior caso.

**ğŸŸ¢ Melhor Caso: O(nÂ²)**
Diferente de outros algoritmos, nÃ£o hÃ¡ otimizaÃ§Ã£o para listas jÃ¡ ordenadas. A complexidade Ã© sempre O(nÂ²).

### ğŸ’¾ EspaÃ§o

**O(1)**
Ordena in-place, usando apenas variÃ¡veis auxiliares para armazenar o Ã­ndice do mÃ­nimo. NÃ£o requer espaÃ§o adicional.

### âš–ï¸ Estabilidade

âŒ NÃ£o - a troca de posiÃ§Ãµes pode alterar a ordem relativa de elementos iguais.

**Exemplo:** [4â‚, 2, 4áµ¦, 1] â†’ [1, 2, 4áµ¦, 4â‚] (4áµ¦ passou na frente de 4â‚)


```python
def selection_sort(arr):
    """
    Sorts an array using selection sort algorithm.
    
    Args:
        arr: List of comparable elements
        
    Returns:
        Sorted list (modified in-place)
        
    Time: O(nÂ²) in all cases
    Space: O(1)
    """
    n = len(arr)
    
    # Outer loop: position to fill with minimum element
    for i in range(n - 1):
        # Assume current position has the minimum
        min_idx = i
        
        # Inner loop: find actual minimum in unsorted part
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        
        # Swap only if minimum is not already in position
        if min_idx != i:
            arr[i], arr[min_idx] = arr[min_idx], arr[i]
    
    return arr
```

## ğŸ’¡ Detalhes da ImplementaÃ§Ã£o

ğŸ”„ O loop externo define a posiÃ§Ã£o que serÃ¡ preenchida com o prÃ³ximo menor elemento. O loop interno varre toda a parte nÃ£o ordenada procurando o Ã­ndice do menor elemento.

ğŸ¯ A variÃ¡vel `min_idx` armazena a posiÃ§Ã£o do menor elemento encontrado. Ao final de cada varredura, fazemos apenas uma troca, colocando o menor na posiÃ§Ã£o correta.

âš¡ A verificaÃ§Ã£o `if min_idx != i` evita trocas desnecessÃ¡rias quando o menor elemento jÃ¡ estÃ¡ na posiÃ§Ã£o correta. Embora nÃ£o melhore a complexidade O(nÂ²), reduz o nÃºmero de escritas na memÃ³ria.


```python
# Test with small array
test_arr = [29, 10, 14, 37, 13]
print(f"Original: {test_arr}")

selection_sort(test_arr)
print(f"Sorted: {test_arr}")
```

```python
# Edge cases
empty = []
single = [42]
sorted_arr = [1, 2, 3, 4, 5]
reverse = [5, 4, 3, 2, 1]
duplicates = [3, 1, 4, 1, 5, 9, 2, 6, 5]

test_cases = [
    ("Empty", empty),
    ("Single element", single),
    ("Already sorted", sorted_arr),
    ("Reverse sorted", reverse),
    ("With duplicates", duplicates)
]

for name, arr in test_cases:
    original = arr.copy()
    selection_sort(arr)
    print(f"{name}: {original} -> {arr}")
```
